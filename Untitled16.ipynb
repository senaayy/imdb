```python
import pandas as pd
import numpy as np
import gradio as gr
import os

print("--- Film √ñneri Sistemi Ba≈ülatƒ±lƒ±yor ---")

# --- ADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor ---
print("ADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor...")

csv_file_name = "imdb-top-rated-movies-user-rated.csv" # Y√ºklediƒüiniz CSV dosyasƒ±nƒ±n adƒ±
file_path = os.path.join("/content/", csv_file_name)

if not os.path.exists(file_path):
    print(f"HATA: '{csv_file_name}' dosyasƒ± '{file_path}' yolunda bulunamadƒ±.")
    print("L√ºtfen CSV dosyasƒ±nƒ± Colab'a doƒüru ≈üekilde y√ºklediƒüinizden ve 'csv_file_name' deƒüi≈ükenini doƒüru ayarladƒ±ƒüƒ±nƒ±zdan emin olun.")
    exit()

df = pd.read_csv(file_path)

print(f"'{csv_file_name}' ba≈üarƒ±yla y√ºklendi. Toplam {len(df)} film bulundu.")
print("\nADIM 1: Veri Seti Ke≈üfi Tamamlandƒ±.")

# --- ADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme ---
print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Ba≈ülƒ±yor (T√ºrler Basitle≈ütiriliyor)...")

df_filtered = df[['Title', 'IMDb Rating', 'Tags', 'Director', 'Stars', 'Votes']].copy()

df_filtered['Stars'].fillna('', inplace=True)

# T√ºrleri Basitle≈ütirme i√ßin e≈üleme s√∂zl√ºƒü√º
genre_mapping = {
    'action': ['action', 'action epic', 'gun fu', 'one-person army action', 'car action', 'kung fu', 'martial arts', 'martial-arts'],
    'adventure': ['adventure', 'adventure epic', 'desert adventure', 'animal adventure', 'space adventure', 'swashbuckler'],
    'comedy': ['comedy', 'romantic comedy', 'buddy comedy', 'sitcom', 'black comedy', 'satire', 'spoof', 'parody', 'slapstick', 'screwball comedy', 'dark comedy', 'body swap comedy'],
    'drama': ['drama', 'period drama', 'cop drama', 'legal drama', 'medical drama', 'teen drama', 'psychological drama', 'melodrama', 'historical drama', 'biography', 'romantic drama', 'showbiz drama', 'tragedy'],
    'thriller': ['thriller', 'crime thriller', 'spy thriller', 'psychological thriller', 'mystery thriller', 'political thriller', 'conspiracy thriller', 'erotic thriller', 'cyber thriller', 'suspense'],
    'sci-fi': ['sci-fi', 'space sci-fi', 'dystopian sci-fi', 'cyberpunk', 'alien invasion', 'mutant', 'robot', 'post-apocalyptic', 'time travel'],
    'fantasy': ['fantasy', 'dark fantasy', 'sword & sorcery', 'fairy tale', 'epic fantasy'],
    'horror': ['horror', 'slasher', 'supernatural horror', 'body horror', 'zombie', 'monster', 'vampire', 'werewolf', 'ghost'],
    'mystery': ['mystery', 'suspense mystery', 'cozy mystery', 'whodunnit', 'detective', 'police procedural'],
    'crime': ['crime', 'gangster', 'heist', 'mob', 'true crime'],
    'romance': ['romance', 'romantic comedy', 'romantic drama'],
    'animation': ['animation', 'adult animation', 'anime', 'computer animation', 'drawn animation', 'stop-motion animation'],
    'family': ['family', 'kids'],
    'western': ['western', 'classic western', 'neo-western'],
    'war': ['war', 'war drama'],
    'history': ['history', 'historical drama', 'biography'],
    'music': ['music', 'musical', 'classic musical', 'concert'],
    'documentary': ['documentary', 'docudrama', 'mockumentary']
}

# Reverse mapping: her bir alt t√ºr√º ana t√ºre e≈üle≈ütirelim
reverse_genre_map = {}
for main_genre, sub_genres in genre_mapping.items():
    for sub_genre in sub_genres:
        reverse_genre_map[sub_genre] = main_genre

def map_to_main_genres(tag_list):
    main_genres = set()
    for tag in tag_list:
        # Alt t√ºr√º ana t√ºre e≈üle, yoksa orijinalini kullan (veya hi√ß ekleme)
        # Orijinalini kullanmak yerine, eƒüer haritalanmamƒ±≈üsa dƒ±≈üarƒ±da bƒ±rakalƒ±m, b√∂ylece daha az g√ºr√ºlt√º olur
        if tag in reverse_genre_map:
            main_genres.add(reverse_genre_map[tag])
    return list(main_genres)


def clean_and_split(text_series):
    # Eƒüer deƒüer zaten NaN ise, bo≈ü liste d√∂nd√ºr
    if pd.isna(text_series):
        return []

    # ensure it's a string before processing
    item = str(text_series)
    item = item.replace('"', '').replace("'", '').strip()
    # "Sci,Fi" gibi hatalƒ± virg√ºlleri d√ºzeltmek i√ßin √∂zel durum
    item = item.replace('sci, fi', 'sci-fi')

    split_items = [s.strip().lower() for s in item.split(',') if s.strip()]
    return split_items


df_filtered['Tags_cleaned_raw'] = df_filtered['Tags'].apply(clean_and_split)
df_filtered['Director_cleaned'] = df_filtered['Director'].apply(clean_and_split)
df_filtered['Stars_cleaned'] = df_filtered['Stars'].apply(clean_and_split)

# Yeni basitle≈ütirilmi≈ü t√ºrler s√ºtunu
df_filtered['Tags_cleaned'] = df_filtered['Tags_cleaned_raw'].apply(map_to_main_genres)


def convert_votes_to_numeric(votes_str):
    if isinstance(votes_str, str):
        votes_str = votes_str.replace(",", "") # Virg√ºlleri temizle (√∂rn. "1,200,000" -> "1200000")
        if 'K' in votes_str:
            return float(votes_str.replace('K', '')) * 1000
        elif 'M' in votes_str:
            return float(votes_str.replace('M', '')) * 1_000_000
    try: # Zaten sayƒ±sal ise veya ba≈üka bir formatta ise direkt d√∂nd√ºr
        return float(votes_str)
    except ValueError:
        return np.nan # D√∂n√º≈üt√ºr√ºlemezse NaN d√∂nd√ºr

df_filtered['Votes_numeric'] = df_filtered['Votes'].apply(convert_votes_to_numeric)
df_filtered.drop('Votes', axis=1, inplace=True)
df_filtered.dropna(subset=['Votes_numeric'], inplace=True) # NaN olanlarƒ± d√º≈ü√ºr

print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Tamamlandƒ±.")


# --- ADIM 3: Film √ñneri Sistemi Mantƒ±ƒüƒ±nƒ± Olu≈üturma ---
print("\nADIM 3: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuluyor...")

# all_tags listesini, sadece e≈ülenmi≈ü ana t√ºrlerden olu≈üturalƒ±m
# Bu sayede Gradio aray√ºz√ºnde √ßok daha temiz ve y√∂netilebilir bir t√ºr listesi olacak.
all_tags = sorted(list(set([tag for sublist in df_filtered['Tags_cleaned'] for tag in sublist if tag in genre_mapping])))

all_directors = sorted(list(set([director for sublist in df_filtered['Director_cleaned'] for director in sublist])))
all_stars = sorted(list(set([star for sublist in df_filtered['Stars_cleaned'] for star in sublist])))


def get_movie_recommendations(selected_tags, selected_directors, selected_stars, min_imdb_rating_slider, num_recommendations_slider):

    # Hata ayƒ±klama √ßƒ±ktƒ±larƒ±
    print(f"\n--- √ñneri ƒ∞steƒüi ---")
    print(f"Se√ßilen T√ºrler (Gradio'dan): {selected_tags}")
    print(f"Se√ßilen Y√∂netmenler (Gradio'dan): {selected_directors}")
    print(f"Se√ßilen Oyuncular (Gradio'dan): {selected_stars}")
    print(f"Minimum IMDb Puanƒ±: {min_imdb_rating_slider}")
    print(f"√ñneri Sayƒ±sƒ±: {num_recommendations_slider}")
    print(f"Ba≈ülangƒ±√ß DataFrame boyutu: {len(df_filtered)}")

    # Gradio'dan gelen se√ßili deƒüerler tuple olarak gelebilir, listeye √ßevirelim
    selected_tags_list = list(selected_tags) if selected_tags else []
    selected_directors_list = list(selected_directors) if selected_directors else []
    selected_stars_list = list(selected_stars) if selected_stars else []

    recommendations_df = df_filtered.copy()

    # 1. IMDb Puanƒ±na G√∂re Filtreleme
    recommendations_df = recommendations_df[recommendations_df['IMDb Rating'] >= min_imdb_rating_slider]
    print(f"IMDb Puanƒ± filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # 2. T√ºrlere G√∂re Filtreleme
    if selected_tags_list:
        recommendations_df = recommendations_df[
            recommendations_df['Tags_cleaned'].apply(lambda x: any(tag in x for tag in selected_tags_list))
        ]
        print(f"T√ºr filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # 3. Y√∂netmenlere G√∂re Filtreleme
    if selected_directors_list:
        recommendations_df = recommendations_df[
            recommendations_df['Director_cleaned'].apply(lambda x: any(director in x for director in selected_directors_list))
        ]
        print(f"Y√∂netmen filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # 4. Oyunculara G√∂re Filtreleme
    if selected_stars_list:
        recommendations_df = recommendations_df[
            recommendations_df['Stars_cleaned'].apply(lambda x: any(star in x for star in selected_stars_list))
        ]
        print(f"Oyuncu filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # 5. Sonu√ßlarƒ± IMDb Puanƒ±na (ve Votes_numeric'e) g√∂re sƒ±ralama
    recommendations_df = recommendations_df.sort_values(
        by=['IMDb Rating', 'Votes_numeric'],
        ascending=[False, False]
    ).reset_index(drop=True)

    # Sadece belirli sayƒ±da √∂neri d√∂nd√ºr (Slider deƒüeri kullanƒ±lƒ±yor)
    top_recommendations = recommendations_df.head(num_recommendations_slider)

    if top_recommendations.empty:
        print("Kriterlere uygun film bulunamadƒ±.")
        return "√úzg√ºn√ºz, se√ßtiƒüiniz kriterlere uygun film bulunamadƒ±."
    else:
        print(f"Toplam {len(top_recommendations)} √∂neri bulundu.")
        recommendation_list = []
        for idx, row in top_recommendations.iterrows():
            directors_str = ", ".join([d.title() for d in row['Director_cleaned']])
            stars_str = ", ".join([s.title() for s in row['Stars_cleaned']])
            tags_str = ", ".join([t.title() for t in row['Tags_cleaned']])

            recommendation_list.append(
                f"**{row['Title']}** (IMDb: {row['IMDb Rating']:.1f} ‚≠êÔ∏è, Oylar: {int(row['Votes_numeric']):,} üó≥Ô∏è)"
                f"\n  Y√∂netmen: {directors_str if directors_str else 'Bilinmiyor'}"
                f"\n  Oyuncular: {stars_str if stars_str else 'Bilinmiyor'}"
                f"\n  T√ºrler: {tags_str if tags_str else 'Bilinmiyor'}"
                f"\n---"
            )
        return "\n".join(recommendation_list)

print("\nADIM 3: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuldu.")


# --- ADIM 4: Gradio Web Aray√ºz√º Olu≈üturma ---
print("\nADIM 4: Gradio Web Aray√ºz√º Olu≈üturuluyor...")

with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown(
        """
        # üé¨ Film √ñneri Sistemi
        Favori film √∂zelliklerinizi se√ßin, y√ºksek IMDb puanƒ±na sahip filmleri ke≈üfedin!
        """
    )

    with gr.Row():
        with gr.Column():
            gr.Markdown("### Film √ñzelliklerini Se√ßin:")

            tags_input = gr.CheckboxGroup(
                label="Film T√ºrleri",
                choices=all_tags,
                value=['action', 'drama'], # Varsayƒ±lan deƒüerler
                interactive=True
            )

            directors_input = gr.Dropdown(
                label="Y√∂netmenler",
                choices=all_directors,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            stars_input = gr.Dropdown(
                label="Oyuncular",
                choices=all_stars,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            min_imdb_rating_slider = gr.Slider(
                minimum=df_filtered['IMDb Rating'].min(),
                maximum=df_filtered['IMDb Rating'].max(),
                step=0.1,
                value=7.6, # Ba≈ülangƒ±√ßta en d√º≈ü√ºk puanƒ± varsayƒ±lan yapalƒ±m, b√∂ylece hep sonu√ß gelir
                label="Minimum IMDb Puanƒ±"
            )

            num_recommendations_slider = gr.Slider(
                minimum=1,
                maximum=20,
                step=1,
                value=10, # Varsayƒ±lan √∂neri sayƒ±sƒ±
                label="√ñneri Sayƒ±sƒ±"
            )

            recommend_btn = gr.Button("üöÄ Film √ñnerilerini Getir", variant="primary", size="lg")

        with gr.Column():
            gr.Markdown("### √ñnerilen Filmler:")
            output_text = gr.Markdown(label="√ñnerileriniz burada listelenecektir.", value="Hen√ºz bir √∂neri yapƒ±lmadƒ±. √ñzellikleri se√ßip butona tƒ±klayƒ±n!")

    # Butonun click olayƒ±na fonksiyonu baƒüla
    recommend_btn.click(
        fn=get_movie_recommendations,
        inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider],
        outputs=output_text
    )

    gr.Examples(
        examples=[
            [['action'], [], [], 7.6, 5],
            [['comedy', 'drama'], [], [], 7.8, 3],
            [[], ['christopher nolan'], [], 8.0, 5],
            [[], [], ['leonardo dicaprio'], 7.8, 3],
        ],
        inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider],
        outputs=output_text,
        fn=get_movie_recommendations, # Examples i√ßin de fonksiyonu belirtmek √∂nemli
        label="√ñrnek √ñnerileri Deneyin"
    )

    gr.Markdown(
        """
        ---
        ### ‚ÑπÔ∏è Nasƒ±l Kullanƒ±lƒ±r?
        1.  **Film T√ºrleri, Y√∂netmenler ve Oyuncular** b√∂l√ºmlerinden istediƒüiniz filtreleri se√ßin (birden fazla se√ßim yapabilirsiniz).
        2.  **Minimum IMDb Puanƒ±** ve **√ñneri Sayƒ±sƒ±** √ßubuklarƒ±nƒ± ayarlayƒ±n.
        3.  **"Film √ñnerilerini Getir"** butonuna tƒ±klayƒ±n.
        4.  √ñneriler saƒü panelde g√∂r√ºnecektir!
        """
    )

demo.launch(share=True)
print("\nADIM 4: Gradio Web Aray√ºz√º Ba≈ülatƒ±ldƒ±.")
```

```python
# Kodu √ßalƒ±≈ütƒ±rmadan √∂nce:
# 1. Colab'da yeni bir not defteri a√ßƒ±n (veya mevcut not defterinin t√ºm √ßalƒ±≈üma zamanƒ±nƒ± sƒ±fƒ±rlayƒ±n: √áalƒ±≈üma Zamanƒ± -> √áalƒ±≈üma zamanƒ±nƒ± sƒ±fƒ±rla).
# 2. √áalƒ±≈üma Zamanƒ± -> √áalƒ±≈üma zamanƒ± t√ºr√ºn√º deƒüi≈ütir -> Donanƒ±m hƒ±zlandƒ±rƒ±cƒ±yƒ± "None" olarak ayarlayƒ±n. (GPU gerekmiyor ama CPU daha stabil √ßalƒ±≈üabilir)
# 3. 'imdb-top-rated-movies-user-rated.csv' dosyasƒ±nƒ± sol paneldeki dosya simgesini kullanarak Colab'a y√ºkleyin.

import pandas as pd
import numpy as np
import gradio as gr
import os
from sentence_transformers import SentenceTransformer, util # Sentence Transformers k√ºt√ºphanesi eklendi
import torch # Embedding'leri Torch tens√∂r√º olarak i≈ülemek i√ßin

print("--- Film √ñneri Sistemi Ba≈ülatƒ±lƒ±yor (NLP Entegrasyonu ile) ---")

# --- ADIM 0: Gerekli K√ºt√ºphaneleri Y√ºkleme (Colab'da genellikle zaten y√ºkl√ºd√ºr ama emin olmak i√ßin) ---
# Bu komutlarƒ± ayrƒ± bir h√ºcrede √ßalƒ±≈ütƒ±rƒ±rsanƒ±z daha iyi olur, ancak tek h√ºcrede de sorun √ßƒ±karmaz.
# Bu kod bloƒüu, sadece tek h√ºcrede √ßalƒ±≈üacak ≈üekilde ayarlandƒ±ƒüƒ± i√ßin burada bƒ±rakƒ±lmƒ±≈ütƒ±r.
try:
    import sentence_transformers
except ImportError:
    print("sentence-transformers k√ºt√ºphanesi y√ºkl√º deƒüil, y√ºkleniyor...")
    !pip install sentence-transformers # Colab'da ! ile terminal komutu √ßalƒ±≈ütƒ±rƒ±rƒ±z
    from sentence_transformers import SentenceTransformer, util
    print("sentence-transformers y√ºklendi.")

# --- ADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor ---
print("\nADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor...")

csv_file_name = "imdb-top-rated-movies-user-rated.csv" # Y√ºklediƒüiniz CSV dosyasƒ±nƒ±n adƒ±
file_path = os.path.join("/content/", csv_file_name)

if not os.path.exists(file_path):
    print(f"HATA: '{csv_file_name}' dosyasƒ± '{file_path}' yolunda bulunamadƒ±.")
    print("L√ºtfen CSV dosyasƒ±nƒ± Colab'a doƒüru ≈üekilde y√ºklediƒüinizden ve 'csv_file_name' deƒüi≈ükenini doƒüru ayarladƒ±ƒüƒ±nƒ±zdan emin olun.")
    exit()

df = pd.read_csv(file_path)

print(f"'{csv_file_name}' ba≈üarƒ±yla y√ºklendi. Toplam {len(df)} film bulundu.")
print("ADIM 1: Veri Seti Ke≈üfi Tamamlandƒ±.")


# --- ADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme ---
print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Ba≈ülƒ±yor (NLP Entegrasyonu i√ßin Hazƒ±rlƒ±k)...")

df_filtered = df[['Title', 'IMDb Rating', 'Tags', 'Director', 'Stars', 'Votes', 'Description']].copy() # 'Description' s√ºtunu eklendi

df_filtered['Stars'].fillna('', inplace=True)
df_filtered['Description'].fillna('', inplace=True) # Description bo≈ü olan filmler i√ßin

# T√ºrleri Basitle≈ütirme i√ßin e≈üleme s√∂zl√ºƒü√º (√ñncekiyle aynƒ±)
genre_mapping = {
    'action': ['action', 'action epic', 'gun fu', 'one-person army action', 'car action', 'kung fu', 'martial arts', 'martial-arts'],
    'adventure': ['adventure', 'adventure epic', 'desert adventure', 'animal adventure', 'space adventure', 'swashbuckler'],
    'comedy': ['comedy', 'romantic comedy', 'buddy comedy', 'sitcom', 'black comedy', 'satire', 'spoof', 'parody', 'slapstick', 'screwball comedy', 'dark comedy', 'body swap comedy'],
    'drama': ['drama', 'period drama', 'cop drama', 'legal drama', 'medical drama', 'teen drama', 'psychological drama', 'melodrama', 'historical drama', 'biography', 'romantic drama', 'showbiz drama', 'tragedy'],
    'thriller': ['thriller', 'crime thriller', 'spy thriller', 'psychological thriller', 'mystery thriller', 'political thriller', 'conspiracy thriller', 'erotic thriller', 'cyber thriller', 'suspense'],
    'sci-fi': ['sci-fi', 'space sci-fi', 'dystopian sci-fi', 'cyberpunk', 'alien invasion', 'mutant', 'robot', 'post-apocalyptic', 'time travel'],
    'fantasy': ['fantasy', 'dark fantasy', 'sword & sorcery', 'fairy tale', 'epic fantasy'],
    'horror': ['horror', 'slasher', 'supernatural horror', 'body horror', 'zombie', 'monster', 'vampire', 'werewolf', 'ghost'],
    'mystery': ['mystery', 'suspense mystery', 'cozy mystery', 'whodunnit', 'detective', 'police procedural'],
    'crime': ['crime', 'gangster', 'heist', 'mob', 'true crime'],
    'romance': ['romance', 'romantic comedy', 'romantic drama'],
    'animation': ['animation', 'adult animation', 'anime', 'computer animation', 'drawn animation', 'stop-motion animation'],
    'family': ['family', 'kids'],
    'western': ['western', 'classic western', 'neo-western'],
    'war': ['war', 'war drama'],
    'history': ['history', 'historical drama', 'biography'],
    'music': ['music', 'musical', 'classic musical', 'concert'],
    'documentary': ['documentary', 'docudrama', 'mockumentary']
}

reverse_genre_map = {}
for main_genre, sub_genres in genre_mapping.items():
    for sub_genre in sub_genres:
        reverse_genre_map[sub_genre] = main_genre

def map_to_main_genres(tag_list):
    main_genres = set()
    for tag in tag_list:
        if tag in reverse_genre_map:
            main_genres.add(reverse_genre_map[tag])
    return list(main_genres)

def clean_and_split(text_series):
    if pd.isna(text_series):
        return []

    item = str(text_series)
    item = item.replace('"', '').replace("'", '').strip()
    item = item.replace('sci, fi', 'sci-fi')

    split_items = [s.strip().lower() for s in item.split(',') if s.strip()]
    return split_items


df_filtered['Tags_cleaned_raw'] = df_filtered['Tags'].apply(clean_and_split)
df_filtered['Director_cleaned'] = df_filtered['Director'].apply(clean_and_split)
df_filtered['Stars_cleaned'] = df_filtered['Stars'].apply(clean_and_split)

df_filtered['Tags_cleaned'] = df_filtered['Tags_cleaned_raw'].apply(map_to_main_genres)


def convert_votes_to_numeric(votes_str):
    if isinstance(votes_str, str):
        votes_str = votes_str.replace(",", "")
        if 'K' in votes_str:
            return float(votes_str.replace('K', '')) * 1000
        elif 'M' in votes_str:
            return float(votes_str.replace('M', '')) * 1_000_000
    try:
        return float(votes_str)
    except ValueError:
        return np.nan

df_filtered['Votes_numeric'] = df_filtered['Votes'].apply(convert_votes_to_numeric)
df_filtered.drop('Votes', axis=1, inplace=True)
df_filtered.dropna(subset=['Votes_numeric'], inplace=True) # NaN olanlarƒ± d√º≈ü√ºr

# NLP i√ßin filmin a√ßƒ±klama metnini birle≈ütirme
df_filtered['Combined_Text'] = df_filtered['Title'] + ". " + \
                               df_filtered['Description'] + ". " + \
                               df_filtered['Tags_cleaned'].apply(lambda x: ", ".join(x)) + ". " + \
                               df_filtered['Director_cleaned'].apply(lambda x: ", ".join(x)) + ". " + \
                               df_filtered['Stars_cleaned'].apply(lambda x: ", ".join(x))

print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Tamamlandƒ±.")


# --- ADIM 3: NLP Modelini Y√ºkleme ve Film Embedding'lerini Olu≈üturma ---
print("\nADIM 3: NLP Modelini Y√ºkleniyor ve Embedding'ler Olu≈üturuluyor...")

# Model y√ºkleme
# Bu model, metinleri sayƒ±sal vekt√∂rlere d√∂n√º≈üt√ºr√ºr.
model_name = 'sentence-transformers/all-MiniLM-L6-v2'
sentence_model = SentenceTransformer(model_name)

# T√ºm filmlerin Combined_Text s√ºtunundan embedding'lerini olu≈üturma
# Bu i≈ülem biraz zaman alabilir (CPU'da birka√ß dakika)
print(f"'{model_name}' modelinden {len(df_filtered)} film i√ßin embedding'ler olu≈üturuluyor...")
film_embeddings = sentence_model.encode(df_filtered['Combined_Text'].tolist(), convert_to_tensor=True, show_progress_bar=True)
print("Embedding'ler ba≈üarƒ±yla olu≈üturuldu.")

print("\nADIM 3: NLP Modelini Y√ºkleme ve Film Embedding'lerini Olu≈üturma Tamamlandƒ±.")


# --- ADIM 4: Film √ñneri Sistemi Mantƒ±ƒüƒ±nƒ± Olu≈üturma (NLP Benzerliƒüi Eklendi) ---
print("\nADIM 4: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuluyor (NLP ile)...")

all_tags = sorted(list(set([tag for sublist in df_filtered['Tags_cleaned'] for tag in sublist if tag in genre_mapping])))
all_directors = sorted(list(set([director for sublist in df_filtered['Director_cleaned'] for director in sublist])))
all_stars = sorted(list(set([star for sublist in df_filtered['Stars_cleaned'] for star in sublist])))

def get_movie_recommendations(selected_tags, selected_directors, selected_stars, min_imdb_rating_slider, num_recommendations_slider, search_text=""):

    # Hata ayƒ±klama √ßƒ±ktƒ±larƒ±
    print(f"\n--- √ñneri ƒ∞steƒüi ---")
    print(f"Se√ßilen T√ºrler: {selected_tags}")
    print(f"Se√ßilen Y√∂netmenler: {selected_directors}")
    print(f"Se√ßilen Oyuncular: {selected_stars}")
    print(f"Minimum IMDb Puanƒ±: {min_imdb_rating_slider}")
    print(f"√ñneri Sayƒ±sƒ±: {num_recommendations_slider}")
    print(f"Arama Metni: '{search_text}'")
    print(f"Ba≈ülangƒ±√ß DataFrame boyutu: {len(df_filtered)}")

    selected_tags_list = list(selected_tags) if selected_tags else []
    selected_directors_list = list(selected_directors) if selected_directors else []
    selected_stars_list = list(selected_stars) if selected_stars else []

    recommendations_df = df_filtered.copy()

    # 1. IMDb Puanƒ±na G√∂re Filtreleme
    recommendations_df = recommendations_df[recommendations_df['IMDb Rating'] >= min_imdb_rating_slider]
    print(f"IMDb Puanƒ± filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # 2. T√ºrlere G√∂re Filtreleme
    if selected_tags_list:
        recommendations_df = recommendations_df[
            recommendations_df['Tags_cleaned'].apply(lambda x: any(tag in x for tag in selected_tags_list))
        ]
        print(f"T√ºr filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # 3. Y√∂netmenlere G√∂re Filtreleme
    if selected_directors_list:
        recommendations_df = recommendations_df[
            recommendations_df['Director_cleaned'].apply(lambda x: any(director in x for director in selected_directors_list))
        ]
        print(f"Y√∂netmen filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # 4. Oyunculara G√∂re Filtreleme
    if selected_stars_list:
        recommendations_df = recommendations_df[
            recommendations_df['Stars_cleaned'].apply(lambda x: any(star in x for star in selected_stars_list))
        ]
        print(f"Oyuncu filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # --- NLP Benzerlik Hesaplamasƒ± (Sadece arama metni varsa) ---
    if search_text and len(recommendations_df) > 0:
        print(f"'{search_text}' i√ßin NLP benzerlik aramasƒ± yapƒ±lƒ±yor...")

        # Arama metninin embedding'ini olu≈ütur
        query_embedding = sentence_model.encode(search_text, convert_to_tensor=True)

        # Mevcut filtrelenmi≈ü filmlerin embedding'lerini al
        # Dikkat: film_embeddings'in indeksi df_filtered'ƒ±n indeksiyle e≈üle≈ümeli
        # Bu y√ºzden recommendations_df'in index'ini kullanarak doƒüru embedding'leri se√ßiyoruz
        filtered_indices = recommendations_df.index.tolist()
        current_film_embeddings = film_embeddings[filtered_indices]

        # Kosin√ºs benzerliƒüi hesapla
        cosine_scores = util.cos_sim(query_embedding, current_film_embeddings)[0]

        # Benzerlik skorlarƒ±nƒ± DataFrame'e ekle
        recommendations_df['Similarity_Score'] = cosine_scores.cpu().numpy() # GPU'dan CPU'ya ta≈üƒ±

        # Benzerlik skoruna g√∂re sƒ±rala (NLP filtresi aktifse)
        recommendations_df = recommendations_df.sort_values(
            by=['Similarity_Score', 'IMDb Rating', 'Votes_numeric'],
            ascending=[False, False, False] # Benzerlik, IMDb, Oylar azalan
        ).reset_index(drop=True)
        print(f"NLP benzerlik filtrelemesi sonrasƒ±: {len(recommendations_df)} film")


    # 5. Sonu√ßlarƒ± IMDb Puanƒ±na (ve Votes_numeric'e) g√∂re sƒ±ralama (NLP filtresi aktif deƒüilse veya son sƒ±rada)
    # Eƒüer search_text yoksa veya NLP benzerliƒüi zaten kullanƒ±ldƒ±ysa, IMDb/Votes sƒ±ralamasƒ± ge√ßerli olacak
    if not search_text: # Eƒüer arama metni yoksa sadece IMDb ve Votes'a g√∂re sƒ±rala
        recommendations_df = recommendations_df.sort_values(
            by=['IMDb Rating', 'Votes_numeric'],
            ascending=[False, False]
        ).reset_index(drop=True)

    # Sadece belirli sayƒ±da √∂neri d√∂nd√ºr (Slider deƒüeri kullanƒ±lƒ±yor)
    top_recommendations = recommendations_df.head(num_recommendations_slider)

    if top_recommendations.empty:
        print("Kriterlere uygun film bulunamadƒ±.")
        return "√úzg√ºn√ºz, se√ßtiƒüiniz kriterlere uygun film bulunamadƒ±."
    else:
        print(f"Toplam {len(top_recommendations)} √∂neri bulundu.")
        recommendation_list = []
        for idx, row in top_recommendations.iterrows():
            directors_str = ", ".join([d.title() for d in row['Director_cleaned']])
            stars_str = ", ".join([s.title() for s in row['Stars_cleaned']])
            tags_str = ", ".join([t.title() for t in row['Tags_cleaned']])

            similarity_info = ""
            if 'Similarity_Score' in row and search_text:
                similarity_info = f", Benzerlik: {row['Similarity_Score']:.2f}"

            recommendation_list.append(
                f"{row['Title']}** (IMDb: {row['IMDb Rating']:.1f} ‚≠ê, Oylar: {int(row['Votes_numeric']):,} üó≥{similarity_info})"
                f"\n  Y√∂netmen: {directors_str if directors_str else 'Bilinmiyor'}"
                f"\n  Oyuncular: {stars_str if stars_str else 'Bilinmiyor'}"
                f"\n  T√ºrler: {tags_str if tags_str else 'Bilinmiyor'}"
                f"\n---"
            )
        return "\n".join(recommendation_list)

print("\nADIM 4: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuldu (NLP Entegrasyonu ile).")


# --- ADIM 5: Gradio Web Aray√ºz√º Olu≈üturma (Arama √áubuƒüu Eklendi) ---
print("\nADIM 5: Gradio Web Aray√ºz√º Olu≈üturuluyor (NLP Arama √áubuƒüu Eklendi)...")

with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown(
        """
        # üé¨ Film √ñneri Sistemi
        Favori film √∂zelliklerinizi se√ßin, y√ºksek IMDb puanƒ±na sahip filmleri ke≈üfedin!
        ƒ∞stediƒüiniz bir film veya konu hakkƒ±nda yazƒ±n, benzerlerini de bulalƒ±m.
        """
    )

    with gr.Row():
        with gr.Column():
            gr.Markdown("### Film √ñzelliklerini Se√ßin:")

            tags_input = gr.CheckboxGroup(
                label="Film T√ºrleri",
                choices=all_tags,
                value=['action', 'drama'], # Varsayƒ±lan deƒüerler
                interactive=True
            )

            directors_input = gr.Dropdown(
                label="Y√∂netmenler",
                choices=all_directors,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            stars_input = gr.Dropdown(
                label="Oyuncular",
                choices=all_stars,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            min_imdb_rating_slider = gr.Slider(
                minimum=df_filtered['IMDb Rating'].min(),
                maximum=df_filtered['IMDb Rating'].max(),
                step=0.1,
                value=7.6, # Ba≈ülangƒ±√ßta en d√º≈ü√ºk puanƒ± varsayƒ±lan yapalƒ±m
                label="Minimum IMDb Puanƒ±"
            )

            num_recommendations_slider = gr.Slider(
                minimum=1,
                maximum=20,
                step=1,
                value=10, # Varsayƒ±lan √∂neri sayƒ±sƒ±
                label="√ñneri Sayƒ±sƒ±"
            )

            # Yeni arama metni giri≈üi
            search_text_input = gr.Textbox(
                label="Film Adƒ± veya Konu Hakkƒ±nda Ara (NLP Tabanlƒ± Benzerlik)",
                placeholder="√ñrneƒüin: Batman, uzay filmi, zamanda yolculuk..."
            )

            recommend_btn = gr.Button("üöÄ Film √ñnerilerini Getir", variant="primary", size="lg")

        with gr.Column():
            gr.Markdown("### √ñnerilen Filmler:")
            output_text = gr.Markdown(label="√ñnerileriniz burada listelenecektir.", value="Hen√ºz bir √∂neri yapƒ±lmadƒ±. √ñzellikleri se√ßip butona tƒ±klayƒ±n!")

    recommend_btn.click(
        fn=get_movie_recommendations,
        inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider, search_text_input],
        outputs=output_text
    )

    gr.Examples(
        examples=[
            [['action'], [], [], 7.6, 5, ""], # Sadece Aksiyon, arama yok
            [['comedy', 'drama'], [], [], 7.8, 3, ""], # Komedi/Drama, arama yok
            [[], ['christopher nolan'], [], 8.0, 5, ""], # Sadece Christopher Nolan filmleri, arama yok
            [[], [], ['leonardo dicaprio'], 7.8, 3, ""], # Sadece Leonardo DiCaprio filmleri, arama yok
            [[], [], [], 8.0, 5, "kahramanlƒ±k ve bilim kurgu"], # Sadece arama metni
            [['action', 'sci-fi'], [], [], 7.8, 5, "uzaylƒ± istilasƒ± ve ka√ßƒ±≈ü"], # T√ºr ve arama metni
        ],
        inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider, search_text_input],
        outputs=output_text,
        fn=get_movie_recommendations,
        label="√ñrnek √ñnerileri Deneyin"
    )

    gr.Markdown(
        """
        ---
        ### ‚Ñπ Nasƒ±l Kullanƒ±lƒ±r?
        1.  *Film T√ºrleri, Y√∂netmenler ve Oyuncular* b√∂l√ºmlerinden istediƒüiniz filtreleri se√ßin (birden fazla se√ßim yapabilirsiniz).
        2.  *Minimum IMDb Puanƒ±* ve *√ñneri Sayƒ±sƒ±* √ßubuklarƒ±nƒ± ayarlayƒ±n.
        3.  ƒ∞sterseniz *"Film Adƒ± veya Konu Hakkƒ±nda Ara"* kutucuƒüuna bir film adƒ±, konu veya anahtar kelime yazƒ±n.
        4.  *"üöÄ Film √ñnerilerini Getir"* butonuna tƒ±klayƒ±n.
        5.  √ñneriler saƒü panelde g√∂r√ºnecektir!
        """
    )

demo.launch(share=True)
print("\nADIM 5: Gradio Web Aray√ºz√º Ba≈ülatƒ±ldƒ± (NLP Arama Aktif).")
```

```python
# Kodu √ßalƒ±≈ütƒ±rmadan √∂nce:
# 1. Colab'da yeni bir not defteri a√ßƒ±n (veya mevcut not defterinin t√ºm √ßalƒ±≈üma zamanƒ±nƒ± sƒ±fƒ±rlayƒ±n: √áalƒ±≈üma Zamanƒ± -> √áalƒ±≈üma zamanƒ±nƒ± sƒ±fƒ±rla).
# 2. √áalƒ±≈üma Zamanƒ± -> √áalƒ±≈üma zamanƒ± t√ºr√ºn√º deƒüi≈ütir -> Donanƒ±m hƒ±zlandƒ±rƒ±cƒ±yƒ± "None" olarak ayarlayƒ±n. (GPU gerekmiyor ama CPU daha stabil √ßalƒ±≈üabilir)
# 3. 'imdb-top-rated-movies-user-rated.csv' dosyasƒ±nƒ± sol paneldeki dosya simgesini kullanarak Colab'a y√ºkleyin.

import pandas as pd
import numpy as np
import gradio as gr
import os
from sentence_transformers import SentenceTransformer, util
import torch

print("--- Film √ñneri Sistemi Ba≈ülatƒ±lƒ±yor (Poster Entegrasyonu ile) ---")

# --- ADIM 0: Gerekli K√ºt√ºphaneleri Y√ºkleme ---
try:
    import sentence_transformers
except ImportError:
    print("sentence-transformers k√ºt√ºphanesi y√ºkl√º deƒüil, y√ºkleniyor...")
    !pip install sentence-transformers
    from sentence_transformers import SentenceTransformer, util
    print("sentence-transformers y√ºklendi.")

# --- ADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor ---
print("\nADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor...")

csv_file_name = "imdb-top-rated-movies-user-rated.csv"
file_path = os.path.join("/content/", csv_file_name)

if not os.path.exists(file_path):
    print(f"HATA: '{csv_file_name}' dosyasƒ± '{file_path}' yolunda bulunamadƒ±.")
    print("L√ºtfen CSV dosyasƒ±nƒ± Colab'a doƒüru ≈üekilde y√ºklediƒüinizden ve 'csv_file_name' deƒüi≈ükenini doƒüru ayarladƒ±ƒüƒ±nƒ±zdan emin olun.")
    exit()

df = pd.read_csv(file_path)

print(f"'{csv_file_name}' ba≈üarƒ±yla y√ºklendi. Toplam {len(df)} film bulundu.")
print("ADIM 1: Veri Seti Ke≈üfi Tamamlandƒ±.")


# --- ADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme ---
print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Ba≈ülƒ±yor (Poster URL'si Eklendi)...")

# 'Poster URL' s√ºtunu da se√ßildi
df_filtered = df[['Title', 'IMDb Rating', 'Tags', 'Director', 'Stars', 'Votes', 'Description', 'Poster URL']].copy()

df_filtered['Stars'].fillna('', inplace=True)
df_filtered['Description'].fillna('', inplace=True)
df_filtered['Poster URL'].fillna('', inplace=True) # Poster URL'si bo≈ü olanlar i√ßin

# T√ºrleri Basitle≈ütirme i√ßin e≈üleme s√∂zl√ºƒü√º (√ñncekiyle aynƒ±)
genre_mapping = {
    'action': ['action', 'action epic', 'gun fu', 'one-person army action', 'car action', 'kung fu', 'martial arts', 'martial-arts'],
    'adventure': ['adventure', 'adventure epic', 'desert adventure', 'animal adventure', 'space adventure', 'swashbuckler'],
    'comedy': ['comedy', 'romantic comedy', 'buddy comedy', 'sitcom', 'black comedy', 'satire', 'spoof', 'parody', 'slapstick', 'screwball comedy', 'dark comedy', 'body swap comedy'],
    'drama': ['drama', 'period drama', 'cop drama', 'legal drama', 'medical drama', 'teen drama', 'psychological drama', 'melodrama', 'historical drama', 'biography', 'romantic drama', 'showbiz drama', 'tragedy'],
    'thriller': ['thriller', 'crime thriller', 'spy thriller', 'psychological thriller', 'mystery thriller', 'political thriller', 'conspiracy thriller', 'erotic thriller', 'cyber thriller', 'suspense'],
    'sci-fi': ['sci-fi', 'space sci-fi', 'dystopian sci-fi', 'cyberpunk', 'alien invasion', 'mutant', 'robot', 'post-apocalyptic', 'time travel'],
    'fantasy': ['fantasy', 'dark fantasy', 'sword & sorcery', 'fairy tale', 'epic fantasy'],
    'horror': ['horror', 'slasher', 'supernatural horror', 'body horror', 'zombie', 'monster', 'vampire', 'werewolf', 'ghost'],
    'mystery': ['mystery', 'suspense mystery', 'cozy mystery', 'whodunnit', 'detective', 'police procedural'],
    'crime': ['crime', 'gangster', 'heist', 'mob', 'true crime'],
    'romance': ['romance', 'romantic comedy', 'romantic drama'],
    'animation': ['animation', 'adult animation', 'anime', 'computer animation', 'drawn animation', 'stop-motion animation'],
    'family': ['family', 'kids'],
    'western': ['western', 'classic western', 'neo-western'],
    'war': ['war', 'war drama'],
    'history': ['history', 'historical drama', 'biography'],
    'music': ['music', 'musical', 'classic musical', 'concert'],
    'documentary': ['documentary', 'docudrama', 'mockumentary']
}

reverse_genre_map = {}
for main_genre, sub_genres in genre_mapping.items():
    for sub_genre in sub_genres:
        reverse_genre_map[sub_genre] = main_genre

def map_to_main_genres(tag_list):
    main_genres = set()
    for tag in tag_list:
        if tag in reverse_genre_map:
            main_genres.add(reverse_genre_map[tag])
    return list(main_genres)

def clean_and_split(text_series):
    if pd.isna(text_series):
        return []

    item = str(text_series)
    item = item.replace('"', '').replace("'", '').strip()
    item = item.replace('sci, fi', 'sci-fi')

    split_items = [s.strip().lower() for s in item.split(',') if s.strip()]
    return split_items


df_filtered['Tags_cleaned_raw'] = df_filtered['Tags'].apply(clean_and_split)
df_filtered['Director_cleaned'] = df_filtered['Director'].apply(clean_and_split)
df_filtered['Stars_cleaned'] = df_filtered['Stars'].apply(clean_and_split)

df_filtered['Tags_cleaned'] = df_filtered['Tags_cleaned_raw'].apply(map_to_main_genres)


def convert_votes_to_numeric(votes_str):
    if isinstance(votes_str, str):
        votes_str = votes_str.replace(",", "")
        if 'K' in votes_str:
            return float(votes_str.replace('K', '')) * 1000
        elif 'M' in votes_str:
            return float(votes_str.replace('M', '')) * 1_000_000
    try:
        return float(votes_str)
    except ValueError:
        return np.nan

df_filtered['Votes_numeric'] = df_filtered['Votes'].apply(convert_votes_to_numeric)
df_filtered.drop('Votes', axis=1, inplace=True)
df_filtered.dropna(subset=['Votes_numeric'], inplace=True)

# NLP i√ßin filmin a√ßƒ±klama metnini birle≈ütirme
df_filtered['Combined_Text'] = df_filtered['Title'] + ". " + \
                               df_filtered['Description'] + ". " + \
                               df_filtered['Tags_cleaned'].apply(lambda x: ", ".join(x)) + ". " + \
                               df_filtered['Director_cleaned'].apply(lambda x: ", ".join(x)) + ". " + \
                               df_filtered['Stars_cleaned'].apply(lambda x: ", ".join(x))

print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Tamamlandƒ±.")


# --- ADIM 3: NLP Modelini Y√ºkleme ve Film Embedding'lerini Olu≈üturma ---
print("\nADIM 3: NLP Modelini Y√ºkleniyor ve Embedding'ler Olu≈üturuluyor...")

model_name = 'sentence-transformers/all-MiniLM-L6-v2'
sentence_model = SentenceTransformer(model_name)

print(f"'{model_name}' modelinden {len(df_filtered)} film i√ßin embedding'ler olu≈üturuluyor...")
film_embeddings = sentence_model.encode(df_filtered['Combined_Text'].tolist(), convert_to_tensor=True, show_progress_bar=True)
print("Embedding'ler ba≈üarƒ±yla olu≈üturuldu.")

print("\nADIM 3: NLP Modelini Y√ºkleme ve Film Embedding'lerini Olu≈üturma Tamamlandƒ±.")


# --- ADIM 4: Film √ñneri Sistemi Mantƒ±ƒüƒ±nƒ± Olu≈üturma (NLP Benzerliƒüi ve Posterler Eklendi) ---
print("\nADIM 4: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuluyor (NLP ve Posterler ile)...")

all_tags = sorted(list(set([tag for sublist in df_filtered['Tags_cleaned'] for tag in sublist if tag in genre_mapping])))
all_directors = sorted(list(set([director for sublist in df_filtered['Director_cleaned'] for director in sublist])))
all_stars = sorted(list(set([star for sublist in df_filtered['Stars_cleaned'] for star in sublist])))

def get_movie_recommendations(selected_tags, selected_directors, selected_stars, min_imdb_rating_slider, num_recommendations_slider, search_text=""):

    print(f"\n--- √ñneri ƒ∞steƒüi ---")
    print(f"Se√ßilen T√ºrler: {selected_tags}")
    print(f"Se√ßilen Y√∂netmenler: {selected_directors}")
    print(f"Se√ßilen Oyuncular: {selected_stars}")
    print(f"Minimum IMDb Puanƒ±: {min_imdb_rating_slider}")
    print(f"√ñneri Sayƒ±sƒ±: {num_recommendations_slider}")
    print(f"Arama Metni: '{search_text}'")
    print(f"Ba≈ülangƒ±√ß DataFrame boyutu: {len(df_filtered)} (Poster URL'si dahil)")

    selected_tags_list = list(selected_tags) if selected_tags else []
    selected_directors_list = list(selected_directors) if selected_directors else []
    selected_stars_list = list(selected_stars) if selected_stars else []

    recommendations_df = df_filtered.copy()

    recommendations_df = recommendations_df[recommendations_df['IMDb Rating'] >= min_imdb_rating_slider]
    print(f"IMDb Puanƒ± filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    if selected_tags_list:
        recommendations_df = recommendations_df[
            recommendations_df['Tags_cleaned'].apply(lambda x: any(tag in x for tag in selected_tags_list))
        ]
        print(f"T√ºr filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    if selected_directors_list:
        recommendations_df = recommendations_df[
            recommendations_df['Director_cleaned'].apply(lambda x: any(director in x for director in selected_directors_list))
        ]
        print(f"Y√∂netmen filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    if selected_stars_list:
        recommendations_df = recommendations_df[
            recommendations_df['Stars_cleaned'].apply(lambda x: any(star in x for star in selected_stars_list))
        ]
        print(f"Oyuncu filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # --- NLP Benzerlik Hesaplamasƒ± (Sadece arama metni varsa) ---
    if search_text and len(recommendations_df) > 0:
        print(f"'{search_text}' i√ßin NLP benzerlik aramasƒ± yapƒ±lƒ±yor...")

        query_embedding = sentence_model.encode(search_text, convert_to_tensor=True)

        filtered_indices = recommendations_df.index.tolist()
        current_film_embeddings = film_embeddings[filtered_indices]

        cosine_scores = util.cos_sim(query_embedding, current_film_embeddings)[0]

        recommendations_df['Similarity_Score'] = cosine_scores.cpu().numpy()

        recommendations_df = recommendations_df.sort_values(
            by=['Similarity_Score', 'IMDb Rating', 'Votes_numeric'],
            ascending=[False, False, False]
        ).reset_index(drop=True)
        print(f"NLP benzerlik filtrelemesi sonrasƒ±: {len(recommendations_df)} film")


    # 5. Sonu√ßlarƒ± IMDb Puanƒ±na (ve Votes_numeric'e) g√∂re sƒ±ralama (NLP filtresi aktif deƒüilse veya son sƒ±rada)
    if not search_text: # Eƒüer arama metni yoksa sadece IMDb ve Votes'a g√∂re sƒ±rala
        recommendations_df = recommendations_df.sort_values(
            by=['IMDb Rating', 'Votes_numeric'],
            ascending=[False, False]
        ).reset_index(drop=True)

    top_recommendations = recommendations_df.head(num_recommendations_slider)

    if top_recommendations.empty:
        print("Kriterlere uygun film bulunamadƒ±.")
        return "√úzg√ºn√ºz, se√ßtiƒüiniz kriterlere uygun film bulunamadƒ±."
    else:
        print(f"Toplam {len(top_recommendations)} √∂neri bulundu.")
        # HTML √ßƒ±ktƒ±sƒ± olu≈üturmak i√ßin
        html_output = ""
        for idx, row in top_recommendations.iterrows():
            directors_str = ", ".join([d.title() for d in row['Director_cleaned']])
            stars_str = ", ".join([s.title() for s in row['Stars_cleaned']])
            tags_str = ", ".join([t.title() for t in row['Tags_cleaned']])

            similarity_info = ""
            if 'Similarity_Score' in row and search_text:
                similarity_info = f", Benzerlik: {row['Similarity_Score']:.2f}"

            # Poster URL'sini al, eƒüer yoksa veya ge√ßersizse placeholder kullan
            poster_url = row['Poster URL'] if row['Poster URL'] and pd.notna(row['Poster URL']) else "https://via.placeholder.com/150x220?text=Poster+Yok"
            # IMDb poster URL'lerinde bazen kƒ±rƒ±k linkler olabiliyor, g√ºvenli bir fallback URL ekleyebiliriz

            html_output += f"""
            <div style="display: flex; margin-bottom: 20px; border: 1px solid #333; padding: 10px; border-radius: 8px; background-color: #1a1a1a;">
                <div style="flex-shrink: 0; margin-right: 15px;">
                    <img src="{poster_url}" alt="{row['Title']} Poster" style="width: 120px; height: 180px; object-fit: cover; border-radius: 4px;">
                </div>
                <div style="flex-grow: 1;">
                    <h3 style="margin-top: 0px; margin-bottom: 5px; color: #f39c12;">{row['Title']}</h3>
                    <p style="margin-bottom: 5px; color: #eee;">IMDb: <b>{row['IMDb Rating']:.1f}</b> ‚≠ê, Oylar: <b>{int(row['Votes_numeric']):,}</b> üó≥Ô∏è{similarity_info}</p>
                    <p style="margin-bottom: 5px; color: #bbb;">Y√∂netmen: {directors_str if directors_str else 'Bilinmiyor'}</p>
                    <p style="margin-bottom: 5px; color: #bbb;">Oyuncular: {stars_str if stars_str else 'Bilinmiyor'}</p>
                    <p style="margin-bottom: 0px; color: #bbb;">T√ºrler: {tags_str if tags_str else 'Bilinmiyor'}</p>
                </div>
            </div>
            """
        return html_output

print("\nADIM 4: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuldu (NLP ve Posterler ile).")


# --- ADIM 5: Gradio Web Aray√ºz√º Olu≈üturma (Arama √áubuƒüu ve Posterler Eklendi) ---
print("\nADIM 5: Gradio Web Aray√ºz√º Olu≈üturuluyor (NLP Arama √áubuƒüu ve Posterler Eklendi)...")

with gr.Blocks(theme=gr.themes.Soft(), css="""
    /* Custom CSS */
    .gradio-container { max-width: 1200px !important; font-family: 'Segoe UI', sans-serif; }
    h1 { color: #f39c12; text-align: center; }
    h3 { color: #eee; }
    .gr-button.gr-button-primary { background-color: #f39c12 !important; border-color: #f39c12 !important; }
    .gr-button.gr-button-primary:hover { background-color: #e67e22 !important; border-color: #e67e22 !important; }
    .gr-checkbox-group label { color: #ccc; }
    .gr-dropdown, .gr-slider, .gr-textbox { background-color: #2c2c2c; color: #eee; border-color: #555; }
    .gr-dropdown-item { color: #eee; }
""") as demo:
    gr.Markdown(
        """
        # üé¨ Film √ñneri Sistemi
        Favori film √∂zelliklerinizi se√ßin, y√ºksek IMDb puanƒ±na sahip filmleri ke≈üfedin!
        ƒ∞stediƒüiniz bir film veya konu hakkƒ±nda yazƒ±n, benzerlerini de bulalƒ±m.
        """
    )

    with gr.Row():
        with gr.Column(scale=1):
            gr.Markdown("### Film √ñzelliklerini Se√ßin:")

            tags_input = gr.CheckboxGroup(
                label="Film T√ºrleri",
                choices=all_tags,
                value=['action', 'drama'],
                interactive=True
            )

            directors_input = gr.Dropdown(
                label="Y√∂netmenler",
                choices=all_directors,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            stars_input = gr.Dropdown(
                label="Oyuncular",
                choices=all_stars,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            min_imdb_rating_slider = gr.Slider(
                minimum=df_filtered['IMDb Rating'].min(),
                maximum=df_filtered['IMDb Rating'].max(),
                step=0.1,
                value=7.6,
                label="Minimum IMDb Puanƒ±"
            )

            num_recommendations_slider = gr.Slider(
                minimum=1,
                maximum=20,
                step=1,
                value=10,
                label="√ñneri Sayƒ±sƒ±"
            )

            search_text_input = gr.Textbox(
                label="Film Adƒ± veya Konu Hakkƒ±nda Ara (NLP Tabanlƒ± Benzerlik)",
                placeholder="√ñrneƒüin: Batman, uzay filmi, zamanda yolculuk..."
            )

            recommend_btn = gr.Button("üöÄ Film √ñnerilerini Getir", variant="primary", size="lg")

        with gr.Column(scale=2): # √áƒ±ktƒ± s√ºtununu daha geni≈ü yaptƒ±k
            gr.Markdown("### √ñnerilen Filmler:")
            # √áƒ±ktƒ± olarak HTML i√ßeriƒüi g√∂sterecek bir bile≈üen kullanƒ±yoruz
            output_html = gr.HTML(label="√ñnerileriniz burada listelenecektir.", value="<p style='text-align: center; color: #bbb;'>Hen√ºz bir √∂neri yapƒ±lmadƒ±. √ñzellikleri se√ßip butona tƒ±klayƒ±n!</p>")

    recommend_btn.click(
        fn=get_movie_recommendations,
        inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider, search_text_input],
        outputs=output_html # HTML √ßƒ±ktƒ±sƒ±nƒ± buraya y√∂nlendiriyoruz
    )

    gr.Examples(
        examples=[
            [['action'], [], [], 7.6, 5, ""],
            [['comedy', 'drama'], [], [], 7.8, 3, ""],
            [[], ['christopher nolan'], [], 8.0, 5, ""],
            [[], [], ['leonardo dicaprio'], 7.8, 3, ""],
            [[], [], [], 8.0, 5, "kahramanlƒ±k ve bilim kurgu"],
            [['action', 'sci-fi'], [], [], 7.8, 5, "uzaylƒ± istilasƒ± ve ka√ßƒ±≈ü"],
        ],
        inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider, search_text_input],
        outputs=output_html, # Examples i√ßin de output_html'i kullanƒ±yoruz
        fn=get_movie_recommendations,
        label="√ñrnek √ñnerileri Deneyin"
    )

    gr.Markdown(
        """
        ---
        ### ‚ÑπÔ∏è Nasƒ±l Kullanƒ±lƒ±r?
        1.  **Film T√ºrleri, Y√∂netmenler ve Oyuncular** b√∂l√ºmlerinden istediƒüiniz filtreleri se√ßin (birden fazla se√ßim yapabilirsiniz).
        2.  **Minimum IMDb Puanƒ±** ve **√ñneri Sayƒ±sƒ±** √ßubuklarƒ±nƒ± ayarlayƒ±n.
        3.  ƒ∞sterseniz **"Film Adƒ± veya Konu Hakkƒ±nda Ara"** kutucuƒüuna bir film adƒ±, konu veya anahtar kelime yazƒ±n.
        4.  **"üöÄ Film √ñnerilerini Getir"** butonuna tƒ±klayƒ±n.
        5.  √ñneriler saƒü panelde g√∂r√ºnecektir!
        """
    )

demo.launch(share=True)
print("\nADIM 5: Gradio Web Aray√ºz√º Ba≈ülatƒ±ldƒ± (NLP Arama √áubuƒüu ve Posterler Eklendi).")
```

```python
import pandas as pd
import numpy as np
import gradio as gr
import os
from sentence_transformers import SentenceTransformer, util
import torch

print("--- Film √ñneri Sistemi Ba≈ülatƒ±lƒ±yor (Poster URL D√ºzeltmesi ile) ---")

# --- ADIM 0: Gerekli K√ºt√ºphaneleri Y√ºkleme ---
try:
    import sentence_transformers
except ImportError:
    print("sentence-transformers k√ºt√ºphanesi y√ºkl√º deƒüil, y√ºkleniyor...")
    !pip install sentence-transformers
    from sentence_transformers import SentenceTransformer, util
    print("sentence-transformers y√ºklendi.")

# --- ADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor ---
print("\nADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor...")

csv_file_name = "imdb-top-rated-movies-user-rated.csv"
file_path = os.path.join("/content/", csv_file_name)

if not os.path.exists(file_path):
    print(f"HATA: '{csv_file_name}' dosyasƒ± '{file_path}' yolunda bulunamadƒ±.")
    print("L√ºtfen CSV dosyasƒ±nƒ± Colab'a doƒüru ≈üekilde y√ºklediƒüinizden ve 'csv_file_name' deƒüi≈ükenini doƒüru ayarladƒ±ƒüƒ±nƒ±zdan emin olun.")
    exit()

df = pd.read_csv(file_path)

print(f"'{csv_file_name}' ba≈üarƒ±yla y√ºklendi. Toplam {len(df)} film bulundu.")
print("ADIM 1: Veri Seti Ke≈üfi Tamamlandƒ±.")


# --- ADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme ---
print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Ba≈ülƒ±yor (Poster URL'si Eklendi)...")

df_filtered = df[['Title', 'IMDb Rating', 'Tags', 'Director', 'Stars', 'Votes', 'Description', 'Poster URL']].copy()

df_filtered['Stars'].fillna('', inplace=True)
df_filtered['Description'].fillna('', inplace=True)
df_filtered['Poster URL'].fillna('', inplace=True)

# T√ºrleri Basitle≈ütirme i√ßin e≈üleme s√∂zl√ºƒü√º
genre_mapping = {
    'action': ['action', 'action epic', 'gun fu', 'one-person army action', 'car action', 'kung fu', 'martial arts', 'martial-arts'],
    'adventure': ['adventure', 'adventure epic', 'desert adventure', 'animal adventure', 'space adventure', 'swashbuckler'],
    'comedy': ['comedy', 'romantic comedy', 'buddy comedy', 'sitcom', 'black comedy', 'satire', 'spoof', 'parody', 'slapstick', 'screwball comedy', 'dark comedy', 'body swap comedy'],
    'drama': ['drama', 'period drama', 'cop drama', 'legal drama', 'medical drama', 'teen drama', 'psychological drama', 'melodrama', 'historical drama', 'biography', 'romantic drama', 'showbiz drama', 'tragedy'],
    'thriller': ['thriller', 'crime thriller', 'spy thriller', 'psychological thriller', 'mystery thriller', 'political thriller', 'conspiracy thriller', 'erotic thriller', 'cyber thriller', 'suspense'],
    'sci-fi': ['sci-fi', 'space sci-fi', 'dystopian sci-fi', 'cyberpunk', 'alien invasion', 'mutant', 'robot', 'post-apocalyptic', 'time travel'],
    'fantasy': ['fantasy', 'dark fantasy', 'sword & sorcery', 'fairy tale', 'epic fantasy'],
    'horror': ['horror', 'slasher', 'supernatural horror', 'body horror', 'zombie', 'monster', 'vampire', 'werewolf', 'ghost'],
    'mystery': ['mystery', 'suspense mystery', 'cozy mystery', 'whodunnit', 'detective', 'police procedural'],
    'crime': ['crime', 'gangster', 'heist', 'mob', 'true crime'],
    'romance': ['romance', 'romantic comedy', 'romantic drama'],
    'animation': ['animation', 'adult animation', 'anime', 'computer animation', 'drawn animation', 'stop-motion animation'],
    'family': ['family', 'kids'],
    'western': ['western', 'classic western', 'neo-western'],
    'war': ['war', 'war drama'],
    'history': ['history', 'historical drama', 'biography'],
    'music': ['music', 'musical', 'classic musical', 'concert'],
    'documentary': ['documentary', 'docudrama', 'mockumentary']
}

reverse_genre_map = {}
for main_genre, sub_genres in genre_mapping.items():
    for sub_genre in sub_genres:
        reverse_genre_map[sub_genre] = main_genre

def map_to_main_genres(tag_list):
    main_genres = set()
    for tag in tag_list:
        if tag in reverse_genre_map:
            main_genres.add(reverse_genre_map[tag])
    return list(main_genres)

def clean_and_split(text_series):
    if pd.isna(text_series):
        return []

    item = str(text_series)
    item = item.replace('"', '').replace("'", '').strip()
    item = item.replace('sci, fi', 'sci-fi')

    split_items = [s.strip().lower() for s in item.split(',') if s.strip()]
    return split_items


df_filtered['Tags_cleaned_raw'] = df_filtered['Tags'].apply(clean_and_split)
df_filtered['Director_cleaned'] = df_filtered['Director'].apply(clean_and_split)
df_filtered['Stars_cleaned'] = df_filtered['Stars'].apply(clean_and_split)

df_filtered['Tags_cleaned'] = df_filtered['Tags_cleaned_raw'].apply(map_to_main_genres)


def convert_votes_to_numeric(votes_str):
    if isinstance(votes_str, str):
        votes_str = votes_str.replace(",", "")
        if 'K' in votes_str:
            return float(votes_str.replace('K', '')) * 1000
        elif 'M' in votes_str:
            return float(votes_str.replace('M', '')) * 1_000_000
    try:
        return float(votes_str)
    except ValueError:
        return np.nan

df_filtered['Votes_numeric'] = df_filtered['Votes'].apply(convert_votes_to_numeric)
df_filtered.drop('Votes', axis=1, inplace=True)
df_filtered.dropna(subset=['Votes_numeric'], inplace=True)

df_filtered['Combined_Text'] = df_filtered['Title'] + ". " + \
                               df_filtered['Description'] + ". " + \
                               df_filtered['Tags_cleaned'].apply(lambda x: ", ".join(x)) + ". " + \
                               df_filtered['Director_cleaned'].apply(lambda x: ", ".join(x)) + ". " + \
                               df_filtered['Stars_cleaned'].apply(lambda x: ", ".join(x))

print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Tamamlandƒ±.")


# --- ADIM 3: NLP Modelini Y√ºkleme ve Film Embedding'lerini Olu≈üturma ---
print("\nADIM 3: NLP Modelini Y√ºkleniyor ve Embedding'ler Olu≈üturuluyor...")

model_name = 'sentence-transformers/all-MiniLM-L6-v2'
sentence_model = SentenceTransformer(model_name)

print(f"'{model_name}' modelinden {len(df_filtered)} film i√ßin embedding'ler olu≈üturuluyor...")
film_embeddings = sentence_model.encode(df_filtered['Combined_Text'].tolist(), convert_to_tensor=True, show_progress_bar=True)
print("Embedding'ler ba≈üarƒ±yla olu≈üturuldu.")

print("\nADIM 3: NLP Modelini Y√ºkleme ve Film Embedding'lerini Olu≈üturma Tamamlandƒ±.")


# --- ADIM 4: Film √ñneri Sistemi Mantƒ±ƒüƒ±nƒ± Olu≈üturma (NLP Benzerliƒüi ve Posterler Eklendi) ---
print("\nADIM 4: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuluyor (NLP ve Posterler ile)...")

all_tags = sorted(list(set([tag for sublist in df_filtered['Tags_cleaned'] for tag in sublist if tag in genre_mapping])))
all_directors = sorted(list(set([director for sublist in df_filtered['Director_cleaned'] for director in sublist])))
all_stars = sorted(list(set([star for sublist in df_filtered['Stars_cleaned'] for star in sublist])))

def get_movie_recommendations(selected_tags, selected_directors, selected_stars, min_imdb_rating_slider, num_recommendations_slider, search_text=""):

    print(f"\n--- √ñneri ƒ∞steƒüi ---")
    print(f"Se√ßilen T√ºrler: {selected_tags}")
    print(f"Se√ßilen Y√∂netmenler: {selected_directors}")
    print(f"Se√ßilen Oyuncular: {selected_stars}")
    print(f"Minimum IMDb Puanƒ±: {min_imdb_rating_slider}")
    print(f"√ñneri Sayƒ±sƒ±: {num_recommendations_slider}")
    print(f"Arama Metni: '{search_text}'")
    print(f"Ba≈ülangƒ±√ß DataFrame boyutu: {len(df_filtered)} (Poster URL'si dahil)")

    selected_tags_list = list(selected_tags) if selected_tags else []
    selected_directors_list = list(selected_directors) if selected_directors else []
    selected_stars_list = list(selected_stars) if selected_stars else []

    recommendations_df = df_filtered.copy()

    recommendations_df = recommendations_df[recommendations_df['IMDb Rating'] >= min_imdb_rating_slider]
    print(f"IMDb Puanƒ± filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    if selected_tags_list:
        recommendations_df = recommendations_df[
            recommendations_df['Tags_cleaned'].apply(lambda x: any(tag in x for tag in selected_tags_list))
        ]
        print(f"T√ºr filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    if selected_directors_list:
        recommendations_df = recommendations_df[
            recommendations_df['Director_cleaned'].apply(lambda x: any(director in x for director in selected_directors_list))
        ]
        print(f"Y√∂netmen filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    if selected_stars_list:
        recommendations_df = recommendations_df[
            recommendations_df['Stars_cleaned'].apply(lambda x: any(star in x for star in selected_stars_list))
        ]
        print(f"Oyuncu filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # --- NLP Benzerlik Hesaplamasƒ± (Sadece arama metni varsa) ---
    if search_text and len(recommendations_df) > 0:
        print(f"'{search_text}' i√ßin NLP benzerlik aramasƒ± yapƒ±lƒ±yor...")

        query_embedding = sentence_model.encode(search_text, convert_to_tensor=True)

        filtered_indices = recommendations_df.index.tolist()
        current_film_embeddings = film_embeddings[filtered_indices]

        cosine_scores = util.cos_sim(query_embedding, current_film_embeddings)[0]

        recommendations_df['Similarity_Score'] = cosine_scores.cpu().numpy()

        recommendations_df = recommendations_df.sort_values(
            by=['Similarity_Score', 'IMDb Rating', 'Votes_numeric'],
            ascending=[False, False, False]
        ).reset_index(drop=True)
        print(f"NLP benzerlik filtrelemesi sonrasƒ±: {len(recommendations_df)} film")


    # 5. Sonu√ßlarƒ± IMDb Puanƒ±na (ve Votes_numeric'e) g√∂re sƒ±ralama (NLP filtresi aktif deƒüilse veya son sƒ±rada)
    if not search_text: # Eƒüer arama metni yoksa sadece IMDb ve Votes'a g√∂re sƒ±rala
        recommendations_df = recommendations_df.sort_values(
            by=['IMDb Rating', 'Votes_numeric'],
            ascending=[False, False]
        ).reset_index(drop=True)

    top_recommendations = recommendations_df.head(num_recommendations_slider)

    if top_recommendations.empty:
        print("Kriterlere uygun film bulunamadƒ±.")
        return "√úzg√ºn√ºz, se√ßtiƒüiniz kriterlere uygun film bulunamadƒ±."
    else:
        print(f"Toplam {len(top_recommendations)} √∂neri bulundu.")
        # HTML √ßƒ±ktƒ±sƒ± olu≈üturmak i√ßin
        html_output = ""
        for idx, row in top_recommendations.iterrows():
            directors_str = ", ".join([d.title() for d in row['Director_cleaned']])
            stars_str = ", ".join([s.title() for s in row['Stars_cleaned']])
            tags_str = ", ".join([t.title() for t in row['Tags_cleaned']])

            similarity_info = ""
            if 'Similarity_Score' in row and search_text:
                similarity_info = f", Benzerlik: {row['Similarity_Score']:.2f}"

            # --- Poster URL D√úZELTMESƒ∞ BURADA ---
            # Ge√ßersiz veya bo≈ü URL'ler i√ßin daha saƒülam bir yer tutucu g√∂rsel URL'si
            poster_url = row['Poster URL'] if row['Poster URL'] and pd.notna(row['Poster URL']) else "https://via.placeholder.com/120x180?text=Poster+Yok"
            # Ayrƒ±ca, bazƒ± IMDb linkleri doƒürudan g√∂rsel linki deƒüil sayfa linki olabilir.
            # ≈ûu an i√ßin bu ≈üekilde bƒ±rakƒ±yoruz, bu veri setinin Poster URL s√ºtunu doƒürudan g√∂rseli g√∂steriyorsa √ßalƒ±≈üƒ±r.
            # √áalƒ±≈ümƒ±yorsa, bu URL'leri alƒ±p manuel kontrol etmek gerekebilir.

            html_output += f"""
            <div style="display: flex; margin-bottom: 20px; border: 1px solid #333; padding: 10px; border-radius: 8px; background-color: #1a1a1a;">
                <div style="flex-shrink: 0; margin-right: 15px;">
                    <img src="{poster_url}" alt="{row['Title']} Poster" style="width: 120px; height: 180px; object-fit: cover; border-radius: 4px;">
                </div>
                <div style="flex-grow: 1;">
                    <h3 style="margin-top: 0px; margin-bottom: 5px; color: #f39c12;">{row['Title']}</h3>
                    <p style="margin-bottom: 5px; color: #eee;">IMDb: <b>{row['IMDb Rating']:.1f}</b> ‚≠ê, Oylar: <b>{int(row['Votes_numeric']):,}</b> üó≥Ô∏è{similarity_info}</p>
                    <p style="margin-bottom: 5px; color: #bbb;">Y√∂netmen: {directors_str if directors_str else 'Bilinmiyor'}</p>
                    <p style="margin-bottom: 5px; color: #bbb;">Oyuncular: {stars_str if stars_str else 'Bilinmiyor'}</p>
                    <p style="margin-bottom: 0px; color: #bbb;">T√ºrler: {tags_str if tags_str else 'Bilinmiyor'}</p>
                </div>
            </div>
            """
        return html_output

print("\nADIM 4: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuldu (NLP ve Posterler ile).")


# --- ADIM 5: Gradio Web Aray√ºz√º Olu≈üturma (Arama √áubuƒüu ve Posterler Eklendi) ---
print("\nADIM 5: Gradio Web Aray√ºz√º Olu≈üturuluyor (NLP Arama √áubuƒüu ve Posterler Eklendi)...")

with gr.Blocks(theme=gr.themes.Soft(), css="""
    /* Custom CSS */
    .gradio-container { max-width: 1200px !important; font-family: 'Segoe UI', sans-serif; }
    h1 { color: #f39c12; text-align: center; }
    h3 { color: #eee; }
    .gr-button.gr-button-primary { background-color: #f39c12 !important; border-color: #f39c12 !important; }
    .gr-button.gr-button-primary:hover { background-color: #e67e22 !important; border-color: #e67e22 !important; }
    .gr-checkbox-group label { color: #ccc; }
    .gr-dropdown, .gr-slider, .gr-textbox { background-color: #2c2c2c; color: #eee; border-color: #555; }
    .gr-dropdown-item { color: #eee; }
""") as demo:
    gr.Markdown(
        """
        # üé¨ Film √ñneri Sistemi
        Favori film √∂zelliklerinizi se√ßin, y√ºksek IMDb puanƒ±na sahip filmleri ke≈üfedin!
        ƒ∞stediƒüiniz bir film veya konu hakkƒ±nda yazƒ±n, benzerlerini de bulalƒ±m.
        """
    )

    with gr.Row():
        with gr.Column(scale=1):
            gr.Markdown("### Film √ñzelliklerini Se√ßin:")

            tags_input = gr.CheckboxGroup(
                label="Film T√ºrleri",
                choices=all_tags,
                value=['action', 'drama'], # Varsayƒ±lan deƒüerler
                interactive=True
            )

            directors_input = gr.Dropdown(
                label="Y√∂netmenler",
                choices=all_directors,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            stars_input = gr.Dropdown(
                label="Oyuncular",
                choices=all_stars,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            min_imdb_rating_slider = gr.Slider(
                minimum=df_filtered['IMDb Rating'].min(),
                maximum=df_filtered['IMDb Rating'].max(),
                step=0.1,
                value=7.6, # Ba≈ülangƒ±√ßta en d√º≈ü√ºk puanƒ± varsayƒ±lan yapalƒ±m
                label="Minimum IMDb Puanƒ±"
            )

            num_recommendations_slider = gr.Slider(
                minimum=1,
                maximum=20,
                step=1,
                value=10, # Varsayƒ±lan √∂neri sayƒ±sƒ±
                label="√ñneri Sayƒ±sƒ±"
            )

            search_text_input = gr.Textbox(
                label="Film Adƒ± veya Konu Hakkƒ±nda Ara (NLP Tabanlƒ± Benzerlik)",
                placeholder="√ñrneƒüin: Batman, uzay filmi, zamanda yolculuk..."
            )

            recommend_btn = gr.Button("üöÄ Film √ñnerilerini Getir", variant="primary", size="lg")

        with gr.Column(scale=2): # √áƒ±ktƒ± s√ºtununu daha geni≈ü yaptƒ±k
            gr.Markdown("### √ñnerilen Filmler:")
            # √áƒ±ktƒ± olarak HTML i√ßeriƒüi g√∂sterecek bir bile≈üen kullanƒ±yoruz
            output_html = gr.HTML(label="√ñnerileriniz burada listelenecektir.", value="<p style='text-align: center; color: #bbb;'>Hen√ºz bir √∂neri yapƒ±lmadƒ±. √ñzellikleri se√ßip butona tƒ±klayƒ±n!</p>")

    recommend_btn.click(
        fn=get_movie_recommendations,
        inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider, search_text_input],
        outputs=output_html
    )

    gr.Examples(
        examples=[
            [['action'], [], [], 7.6, 5, ""],
            [['comedy', 'drama'], [], [], 7.8, 3, ""],
            [[], ['christopher nolan'], [], 8.0, 5, ""],
            [[], [], ['leonardo dicaprio'], 7.8, 3, ""],
            [[], [], [], 8.0, 5, "kahramanlƒ±k ve bilim kurgu"],
            [['action', 'sci-fi'], [], [], 7.8, 5, "uzaylƒ± istilasƒ± ve ka√ßƒ±≈ü"],
        ],
        inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider, search_text_input],
        outputs=output_html,
        fn=get_movie_recommendations,
        label="√ñrnek √ñnerileri Deneyin"
    )

    gr.Markdown(
        """
        ---
        ### ‚ÑπÔ∏è Nasƒ±l Kullanƒ±lƒ±r?
        1.  **Film T√ºrleri, Y√∂netmenler ve Oyuncular** b√∂l√ºmlerinden istediƒüiniz filtreleri se√ßin (birden fazla se√ßim yapabilirsiniz).
        2.  **Minimum IMDb Puanƒ±** ve **√ñneri Sayƒ±sƒ±** √ßubuklarƒ±nƒ± ayarlayƒ±n.
        3.  ƒ∞sterseniz **"Film Adƒ± veya Konu Hakkƒ±nda Ara"** kutucuƒüuna bir film adƒ±, konu veya anahtar kelime yazƒ±n.
        4.  **"üöÄ Film √ñnerilerini Getir"** butonuna tƒ±klayƒ±n.
        5.  √ñneriler saƒü panelde g√∂r√ºnecektir!
        """
    )

demo.launch(share=True)
print("\nADIM 5: Gradio Web Aray√ºz√º Ba≈ülatƒ±ldƒ± (NLP Arama √áubuƒüu ve Posterler Eklendi).")
```

```python
import pandas as pd
import numpy as np
import gradio as gr
import os
from sentence_transformers import SentenceTransformer, util
import torch

print("--- Film √ñneri Sistemi Ba≈ülatƒ±lƒ±yor (Poster Yer Tutucu ƒ∞yile≈ütirmesi ile) ---")

# --- ADIM 0: Gerekli K√ºt√ºphaneleri Y√ºkleme ---
try:
    import sentence_transformers
except ImportError:
    print("sentence-transformers k√ºt√ºphanesi y√ºkl√º deƒüil, y√ºkleniyor...")
    !pip install sentence-transformers
    from sentence_transformers import SentenceTransformer, util
    print("sentence-transformers y√ºklendi.")

# --- ADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor ---
print("\nADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor...")

csv_file_name = "imdb-top-rated-movies-user-rated.csv"
file_path = os.path.join("/content/", csv_file_name)

if not os.path.exists(file_path):
    print(f"HATA: '{csv_file_name}' dosyasƒ± '{file_path}' yolunda bulunamadƒ±.")
    print("L√ºtfen CSV dosyasƒ±nƒ± Colab'a doƒüru ≈üekilde y√ºklediƒüinizden ve 'csv_file_name' deƒüi≈ükenini doƒüru ayarladƒ±ƒüƒ±nƒ±zdan emin olun.")
    exit()

df = pd.read_csv(file_path)

print(f"'{csv_file_name}' ba≈üarƒ±yla y√ºklendi. Toplam {len(df)} film bulundu.")
print("ADIM 1: Veri Seti Ke≈üfi Tamamlandƒ±.")


# --- ADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme ---
print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Ba≈ülƒ±yor (Poster URL'si Eklendi)...")

df_filtered = df[['Title', 'IMDb Rating', 'Tags', 'Director', 'Stars', 'Votes', 'Description', 'Poster URL']].copy()

df_filtered['Stars'].fillna('', inplace=True)
df_filtered['Description'].fillna('', inplace=True)
df_filtered['Poster URL'].fillna('', inplace=True)

# T√ºrleri Basitle≈ütirme i√ßin e≈üleme s√∂zl√ºƒü√º
genre_mapping = {
    'action': ['action', 'action epic', 'gun fu', 'one-person army action', 'car action', 'kung fu', 'martial arts', 'martial-arts'],
    'adventure': ['adventure', 'adventure epic', 'desert adventure', 'animal adventure', 'space adventure', 'swashbuckler'],
    'comedy': ['comedy', 'romantic comedy', 'buddy comedy', 'sitcom', 'black comedy', 'satire', 'spoof', 'parody', 'slapstick', 'screwball comedy', 'dark comedy', 'body swap comedy'],
    'drama': ['drama', 'period drama', 'cop drama', 'legal drama', 'medical drama', 'teen drama', 'psychological drama', 'melodrama', 'historical drama', 'biography', 'romantic drama', 'showbiz drama', 'tragedy'],
    'thriller': ['thriller', 'crime thriller', 'spy thriller', 'psychological thriller', 'mystery thriller', 'political thriller', 'conspiracy thriller', 'erotic thriller', 'cyber thriller', 'suspense'],
    'sci-fi': ['sci-fi', 'space sci-fi', 'dystopian sci-fi', 'cyberpunk', 'alien invasion', 'mutant', 'robot', 'post-apocalyptic', 'time travel'],
    'fantasy': ['fantasy', 'dark fantasy', 'sword & sorcery', 'fairy tale', 'epic fantasy'],
    'horror': ['horror', 'slasher', 'supernatural horror', 'body horror', 'zombie', 'monster', 'vampire', 'werewolf', 'ghost'],
    'mystery': ['mystery', 'suspense mystery', 'cozy mystery', 'whodunnit', 'detective', 'police procedural'],
    'crime': ['crime', 'gangster', 'heist', 'mob', 'true crime'],
    'romance': ['romance', 'romantic comedy', 'romantic drama'],
    'animation': ['animation', 'adult animation', 'anime', 'computer animation', 'drawn animation', 'stop-motion animation'],
    'family': ['family', 'kids'],
    'western': ['western', 'classic western', 'neo-western'],
    'war': ['war', 'war drama'],
    'history': ['history', 'historical drama', 'biography'],
    'music': ['music', 'musical', 'classic musical', 'concert'],
    'documentary': ['documentary', 'docudrama', 'mockumentary']
}

reverse_genre_map = {}
for main_genre, sub_genres in genre_mapping.items():
    for sub_genre in sub_genres:
        reverse_genre_map[sub_genre] = main_genre

def map_to_main_genres(tag_list):
    main_genres = set()
    for tag in tag_list:
        if tag in reverse_genre_map:
            main_genres.add(reverse_genre_map[tag])
    return list(main_genres)

def clean_and_split(text_series):
    if pd.isna(text_series):
        return []

    item = str(text_series)
    item = item.replace('"', '').replace("'", '').strip()
    item = item.replace('sci, fi', 'sci-fi')

    split_items = [s.strip().lower() for s in item.split(',') if s.strip()]
    return split_items


df_filtered['Tags_cleaned_raw'] = df_filtered['Tags'].apply(clean_and_split)
df_filtered['Director_cleaned'] = df_filtered['Director'].apply(clean_and_split)
df_filtered['Stars_cleaned'] = df_filtered['Stars'].apply(clean_and_split)

df_filtered['Tags_cleaned'] = df_filtered['Tags_cleaned_raw'].apply(map_to_main_genres)


def convert_votes_to_numeric(votes_str):
    if isinstance(votes_str, str):
        votes_str = votes_str.replace(",", "")
        if 'K' in votes_str:
            return float(votes_str.replace('K', '')) * 1000
        elif 'M' in votes_str:
            return float(votes_str.replace('M', '')) * 1_000_000
    try:
        return float(votes_str)
    except ValueError:
        return np.nan

df_filtered['Votes_numeric'] = df_filtered['Votes'].apply(convert_votes_to_numeric)
df_filtered.drop('Votes', axis=1, inplace=True)
df_filtered.dropna(subset=['Votes_numeric'], inplace=True)

df_filtered['Combined_Text'] = df_filtered['Title'] + ". " + \
                               df_filtered['Description'] + ". " + \
                               df_filtered['Tags_cleaned'].apply(lambda x: ", ".join(x)) + ". " + \
                               df_filtered['Director_cleaned'].apply(lambda x: ", ".join(x)) + ". " + \
                               df_filtered['Stars_cleaned'].apply(lambda x: ", ".join(x))

print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Tamamlandƒ±.")


# --- ADIM 3: NLP Modelini Y√ºkleme ve Film Embedding'lerini Olu≈üturma ---
print("\nADIM 3: NLP Modelini Y√ºkleniyor ve Embedding'ler Olu≈üturuluyor...")

model_name = 'sentence-transformers/all-MiniLM-L6-v2'
sentence_model = SentenceTransformer(model_name)

print(f"'{model_name}' modelinden {len(df_filtered)} film i√ßin embedding'ler olu≈üturuluyor...")
film_embeddings = sentence_model.encode(df_filtered['Combined_Text'].tolist(), convert_to_tensor=True, show_progress_bar=True)
print("Embedding'ler ba≈üarƒ±yla olu≈üturuldu.")

print("\nADIM 3: NLP Modelini Y√ºkleme ve Film Embedding'lerini Olu≈üturma Tamamlandƒ±.")


# --- ADIM 4: Film √ñneri Sistemi Mantƒ±ƒüƒ±nƒ± Olu≈üturma (NLP Benzerliƒüi ve Posterler Eklendi) ---
print("\nADIM 4: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuluyor (NLP ve Posterler ile)...")

all_tags = sorted(list(set([tag for sublist in df_filtered['Tags_cleaned'] for tag in sublist if tag in genre_mapping])))
all_directors = sorted(list(set([director for sublist in df_filtered['Director_cleaned'] for director in sublist])))
all_stars = sorted(list(set([star for sublist in df_filtered['Stars_cleaned'] for star in sublist])))

def get_movie_recommendations(selected_tags, selected_directors, selected_stars, min_imdb_rating_slider, num_recommendations_slider, search_text=""):

    print(f"\n--- √ñneri ƒ∞steƒüi ---")
    print(f"Se√ßilen T√ºrler: {selected_tags}")
    print(f"Se√ßilen Y√∂netmenler: {selected_directors}")
    print(f"Se√ßilen Oyuncular: {selected_stars}")
    print(f"Minimum IMDb Puanƒ±: {min_imdb_rating_slider}")
    print(f"√ñneri Sayƒ±sƒ±: {num_recommendations_slider}")
    print(f"Arama Metni: '{search_text}'")
    print(f"Ba≈ülangƒ±√ß DataFrame boyutu: {len(df_filtered)} (Poster URL'si dahil)")

    selected_tags_list = list(selected_tags) if selected_tags else []
    selected_directors_list = list(selected_directors) if selected_directors else []
    selected_stars_list = list(selected_stars) if selected_stars else []

    recommendations_df = df_filtered.copy()

    recommendations_df = recommendations_df[recommendations_df['IMDb Rating'] >= min_imdb_rating_slider]
    print(f"IMDb Puanƒ± filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    if selected_tags_list:
        recommendations_df = recommendations_df[
            recommendations_df['Tags_cleaned'].apply(lambda x: any(tag in x for tag in selected_tags_list))
        ]
        print(f"T√ºr filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    if selected_directors_list:
        recommendations_df = recommendations_df[
            recommendations_df['Director_cleaned'].apply(lambda x: any(director in x for director in selected_directors_list))
        ]
        print(f"Y√∂netmen filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    if selected_stars_list:
        recommendations_df = recommendations_df[
            recommendations_df['Stars_cleaned'].apply(lambda x: any(star in x for star in selected_stars_list))
        ]
        print(f"Oyuncu filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # --- NLP Benzerlik Hesaplamasƒ± (Sadece arama metni varsa) ---
    if search_text and len(recommendations_df) > 0:
        print(f"'{search_text}' i√ßin NLP benzerlik aramasƒ± yapƒ±lƒ±yor...")

        query_embedding = sentence_model.encode(search_text, convert_to_tensor=True)

        filtered_indices = recommendations_df.index.tolist()
        current_film_embeddings = film_embeddings[filtered_indices]

        cosine_scores = util.cos_sim(query_embedding, current_film_embeddings)[0]

        recommendations_df['Similarity_Score'] = cosine_scores.cpu().numpy()

        recommendations_df = recommendations_df.sort_values(
            by=['Similarity_Score', 'IMDb Rating', 'Votes_numeric'],
            ascending=[False, False, False]
        ).reset_index(drop=True)
        print(f"NLP benzerlik filtrelemesi sonrasƒ±: {len(recommendations_df)} film")


    # 5. Sonu√ßlarƒ± IMDb Puanƒ±na (ve Votes_numeric'e) g√∂re sƒ±ralama (NLP filtresi aktif deƒüilse veya son sƒ±rada)
    if not search_text: # Eƒüer arama metni yoksa sadece IMDb ve Votes'a g√∂re sƒ±rala
        recommendations_df = recommendations_df.sort_values(
            by=['IMDb Rating', 'Votes_numeric'],
            ascending=[False, False]
        ).reset_index(drop=True)

    top_recommendations = recommendations_df.head(num_recommendations_slider)

    if top_recommendations.empty:
        print("Kriterlere uygun film bulunamadƒ±.")
        return "√úzg√ºn√ºz, se√ßtiƒüiniz kriterlere uygun film bulunamadƒ±."
    else:
        print(f"Toplam {len(top_recommendations)} √∂neri bulundu.")
        # HTML √ßƒ±ktƒ±sƒ± olu≈üturmak i√ßin
        html_output = ""
        for idx, row in top_recommendations.iterrows():
            directors_str = ", ".join([d.title() for d in row['Director_cleaned']])
            stars_str = ", ".join([s.title() for s in row['Stars_cleaned']])
            tags_str = ", ".join([t.title() for t in row['Tags_cleaned']])

            similarity_info = ""
            if 'Similarity_Score' in row and search_text:
                similarity_info = f", Benzerlik: {row['Similarity_Score']:.2f}"

            # --- Poster URL D√úZELTMESƒ∞ BURADA ---
            # Ge√ßersiz veya bo≈ü URL'ler i√ßin daha saƒülam ve CSS ile kontrol edilebilir bir yer tutucu
            poster_url = row['Poster URL'] # Orijinal URL'yi alalƒ±m

            # Ge√ßerli bir g√∂rsel URL'si olup olmadƒ±ƒüƒ±nƒ± kontrol eden k√º√ß√ºk bir fonksiyon
            # Bu, tarayƒ±cƒ±nƒ±n kƒ±rƒ±k g√∂rsel ikonu g√∂stermesini engeller.
            # Ancak, Gradio'da bu kontrol√º doƒürudan yapamayƒ±z, tarayƒ±cƒ±ya bƒ±rakƒ±rƒ±z.
            # En iyi y√∂ntem, ge√ßerli bir URL yoksa doƒürudan placeholder CSS'ini uygulamaktƒ±r.

            # Buradaki HTML yapƒ±sƒ±nƒ± deƒüi≈ütirelim: Poster URL'si yoksa sadece isim ve a√ßƒ±klama olsun
            # ve poster i√ßin bir yer tutucu divi olsun.

            # Kontrol: URL ger√ßekten bir g√∂rseli i≈üaret ediyor mu? Basit bir regex kontrol√º
            import re
            is_valid_image_url = False
            if poster_url and pd.notna(poster_url):
                # .jpg, .png, .jpeg gibi uzantƒ±larƒ± kontrol et
                if re.match(r".*\.(jpg|jpeg|png|gif|bmp|webp)$", poster_url.lower()):
                    is_valid_image_url = True

            poster_html = ""
            if is_valid_image_url:
                 poster_html = f'<img src="{poster_url}" alt="{row["Title"]} Poster" style="width: 120px; height: 180px; object-fit: cover; border-radius: 4px;">'
            else:
                # Poster URL'si yoksa veya ge√ßersizse bir yer tutucu divi g√∂ster
                poster_html = f"""
                <div style="width: 120px; height: 180px; background-color: #2a2a2a; border-radius: 4px; display: flex; align-items: center; justify-content: center; text-align: center; color: #888; font-size: 0.8em; line-height: 1.2;">
                    {row['Title']} <br>Poster Yok
                </div>
                """


            html_output += f"""
            <div style="display: flex; margin-bottom: 20px; border: 1px solid #333; padding: 10px; border-radius: 8px; background-color: #1a1a1a;">
                <div style="flex-shrink: 0; margin-right: 15px;">
                    {poster_html}
                </div>
                <div style="flex-grow: 1;">
                    <h3 style="margin-top: 0px; margin-bottom: 5px; color: #f39c12;">{row['Title']}</h3>
                    <p style="margin-bottom: 5px; color: #eee;">IMDb: <b>{row['IMDb Rating']:.1f}</b> ‚≠ê, Oylar: <b>{int(row['Votes_numeric']):,}</b> üó≥Ô∏è{similarity_info}</p>
                    <p style="margin-bottom: 5px; color: #bbb;">Y√∂netmen: {directors_str if directors_str else 'Bilinmiyor'}</p>
                    <p style="margin-bottom: 5px; color: #bbb;">Oyuncular: {stars_str if stars_str else 'Bilinmiyor'}</p>
                    <p style="margin-bottom: 0px; color: #bbb;">T√ºrler: {tags_str if tags_str else 'Bilinmiyor'}</p>
                </div>
            </div>
            """
        return html_output

print("\nADIM 4: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuldu (NLP ve Posterler ile).")


# --- ADIM 5: Gradio Web Aray√ºz√º Olu≈üturma (Arama √áubuƒüu ve Posterler Eklendi) ---
print("\nADIM 5: Gradio Web Aray√ºz√º Olu≈üturuluyor (NLP Arama √áubuƒüu ve Posterler Eklendi)...")

with gr.Blocks(theme=gr.themes.Soft(), css="""
    /* Custom CSS */
    .gradio-container { max-width: 1200px !important; font-family: 'Segoe UI', sans-serif; }
    h1 { color: #f39c12; text-align: center; }
    h3 { color: #eee; }
    .gr-button.gr-button-primary { background-color: #f39c12 !important; border-color: #f39c12 !important; }
    .gr-button.gr-button-primary:hover { background-color: #e67e22 !important; border-color: #e67e22 !important; }
    .gr-checkbox-group label { color: #ccc; }
    .gr-dropdown, .gr-slider, .gr-textbox { background-color: #2c2c2c; color: #eee; border-color: #555; }
    .gr-dropdown-item { color: #eee; }
""") as demo:
    gr.Markdown(
        """
        # üé¨ Film √ñneri Sistemi
        Favori film √∂zelliklerinizi se√ßin, y√ºksek IMDb puanƒ±na sahip filmleri ke≈üfedin!
        ƒ∞stediƒüiniz bir film veya konu hakkƒ±nda yazƒ±n, benzerlerini de bulalƒ±m.
        """
    )

    with gr.Row():
        with gr.Column(scale=1):
            gr.Markdown("### Film √ñzelliklerini Se√ßin:")

            tags_input = gr.CheckboxGroup(
                label="Film T√ºrleri",
                choices=all_tags,
                value=['action', 'drama'], # Varsayƒ±lan deƒüerler
                interactive=True
            )

            directors_input = gr.Dropdown(
                label="Y√∂netmenler",
                choices=all_directors,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            stars_input = gr.Dropdown(
                label="Oyuncular",
                choices=all_stars,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            min_imdb_rating_slider = gr.Slider(
                minimum=df_filtered['IMDb Rating'].min(),
                maximum=df_filtered['IMDb Rating'].max(),
                step=0.1,
                value=7.6,
                label="Minimum IMDb Puanƒ±"
            )

            num_recommendations_slider = gr.Slider(
                minimum=1,
                maximum=20,
                step=1,
                value=10,
                label="√ñneri Sayƒ±sƒ±"
            )

            search_text_input = gr.Textbox(
                label="Film Adƒ± veya Konu Hakkƒ±nda Ara (NLP Tabanlƒ± Benzerlik)",
                placeholder="√ñrneƒüin: Batman, uzay filmi, zamanda yolculuk..."
            )

            recommend_btn = gr.Button("üöÄ Film √ñnerilerini Getir", variant="primary", size="lg")

        with gr.Column(scale=2):
            gr.Markdown("### √ñnerilen Filmler:")
            output_html = gr.HTML(label="√ñnerileriniz burada listelenecektir.", value="<p style='text-align: center; color: #bbb;'>Hen√ºz bir √∂neri yapƒ±lmadƒ±. √ñzellikleri se√ßip butona tƒ±klayƒ±n!</p>")

    recommend_btn.click(
        fn=get_movie_recommendations,
        inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider, search_text_input],
        outputs=output_html
    )

    gr.Examples(
        examples=[
            [['action'], [], [], 7.6, 5, ""],
            [['comedy', 'drama'], [], [], 7.8, 3, ""],
            [[], ['christopher nolan'], [], 8.0, 5, ""],
            [[], [], ['leonardo dicaprio'], 7.8, 3, ""],
            [[], [], [], 8.0, 5, "kahramanlƒ±k ve bilim kurgu"],
            [['action', 'sci-fi'], [], [], 7.8, 5, "uzaylƒ± istilasƒ± ve ka√ßƒ±≈ü"],
        ],
        inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider, search_text_input],
        outputs=output_html,
        fn=get_movie_recommendations,
        label="√ñrnek √ñnerileri Deneyin"
    )

    gr.Markdown(
        """
        ---
        ### ‚ÑπÔ∏è Nasƒ±l Kullanƒ±lƒ±r?
        1.  **Film T√ºrleri, Y√∂netmenler ve Oyuncular** b√∂l√ºmlerinden istediƒüiniz filtreleri se√ßin (birden fazla se√ßim yapabilirsiniz).
        2.  **Minimum IMDb Puanƒ±** ve **√ñneri Sayƒ±sƒ±** √ßubuklarƒ±nƒ± ayarlayƒ±n.
        3.  ƒ∞sterseniz **"Film Adƒ± veya Konu Hakkƒ±nda Ara"** kutucuƒüuna bir film adƒ±, konu veya anahtar kelime yazƒ±n.
        4.  **"üöÄ Film √ñnerilerini Getir"** butonuna tƒ±klayƒ±n.
        5.  √ñneriler saƒü panelde g√∂r√ºnecektir!
        """
    )

demo.launch(share=True)
print("\nADIM 5: Gradio Web Aray√ºz√º Ba≈ülatƒ±ldƒ± (NLP Arama √áubuƒüu ve Posterler Eklendi).")
```
