```python
import pandas as pd
import numpy as np
import gradio as gr
import os

print("ADIM 1: Veri Seti Yükleniyor ve Keşfediliyor...")

# Yüklediğiniz CSV dosyasının adını buraya yazın
# Ekran görüntünüzden dosya adının 'imdb-top-rated-movies-user-rated.csv' olduğunu görüyorum.
csv_file_name = "imdb-top-rated-movies-user-rated.csv"

# Colab'daki varsayılan yükleme konumu /content/ olduğundan
file_path = os.path.join("/content/", csv_file_name)

if not os.path.exists(file_path):
    print(f"HATA: '{csv_file_name}' dosyası '{file_path}' yolunda bulunamadı.")
    print("Lütfen CSV dosyasını Colab'a doğru şekilde yüklediğinizden ve 'csv_file_name' değişkenini doğru ayarladığınızdan emin olun.")
    exit()

df = pd.read_csv(file_path)

print(f"'{csv_file_name}' başarıyla yüklendi. Toplam {len(df)} film bulundu.")
print("\nVeri Setinin İlk 5 Satırı:")
print(df.head())

print("\nVeri Setindeki Sütunlar ve Veri Tipleri:")
print(df.info())

print("\nIMDb Puanı Dağılımı:")
# Ekran görüntünüzdeki sütun başlığı 'IMDb Rating' (Boşluklu ve büyük harfli) olduğu için onu kullanıyoruz.
print(df['IMDb Rating'].describe())

print("\nADIM 1: Veri Seti Keşfi Tamamlandı.")
```

```python
import pandas as pd
import numpy as np

print("ADIM 2: Veri Temizliği ve Ön İşleme Başlıyor...")

# ADIM 1'den df DataFrame'ini kullanıyoruz

# 1. İhtiyacımız Olan Sütunları Seçme
# Öneri sistemi için kullanacağımız sütunlar:
# Title (Başlık), IMDb Rating (IMDb Puanı), Tags (Türler), Director (Yönetmen), Stars (Oyuncular), Votes (Oy Sayısı)
df_filtered = df[['Title', 'IMDb Rating', 'Tags', 'Director', 'Stars', 'Votes']].copy()

# 2. Eksik Verileri İşleme (NaN değerler)
# IMDb Rating, Tags, Director sütunlarında eksik veri olmamalı (df.info() çıktısına göre yok)
# Stars sütununda 1 eksik değer vardı. Bu tür küçük eksiklikleri boş string ile doldurabiliriz.
df_filtered['Stars'].fillna('', inplace=True)

# Eksik değerleri doldurduktan sonra tekrar kontrol edelim
print("\nEksik değer kontrolü (fillna sonrası):")
print(df_filtered.isnull().sum())

# 3. 'Tags', 'Director', 'Stars' Sütunlarını Standartlaştırma ve Temizleme
# Bu sütunlar genellikle virgülle ayrılmış birden fazla değeri içerir (örn. "Action, Drama")
# Bazıları tırnak içinde olabilir, onları da temizleyelim.

def clean_and_split(text_series):
    # NaN ise boş liste döndür
    if text_series.isnull().any():
        text_series = text_series.fillna('')

    # Her bir öğeyi temizle: tırnakları kaldır, boşlukları düzelt, küçük harfe çevir
    cleaned_list = []
    for item in text_series.astype(str): # Hata vermemesi için stringe çevir
        item = item.replace('"', '').replace("'", '').strip() # Tırnakları ve gereksiz boşlukları kaldır
        split_items = [s.strip().lower() for s in item.split(',') if s.strip()] # Virgülle ayır, boşlukları düzelt, küçük harfe çevir
        cleaned_list.append(split_items)
    return cleaned_list

df_filtered['Tags_cleaned'] = clean_and_split(df_filtered['Tags'])
df_filtered['Director_cleaned'] = clean_and_split(df_filtered['Director'])
df_filtered['Stars_cleaned'] = clean_and_split(df_filtered['Stars'])

# 4. 'Votes' Sütununu Sayısal Değere Çevirme
# "927K" gibi değerler var. 'K' varsa 1000 ile çarpacağız.
def convert_votes_to_numeric(votes_str):
    if isinstance(votes_str, str):
        if 'K' in votes_str:
            return float(votes_str.replace('K', '')) * 1000
        elif 'M' in votes_str: # Milyon için de eklendi
            return float(votes_str.replace('M', '')) * 1_000_000
    return float(votes_str) # Zaten sayısal ise veya başka bir formatta ise direkt çevir

df_filtered['Votes_numeric'] = df_filtered['Votes'].apply(convert_votes_to_numeric)

# Gereksiz Votes sütununu kaldır
df_filtered.drop('Votes', axis=1, inplace=True)


print("\nTemizlenmiş Veri Setinin İlk 5 Satırı (Yeni Sütunlarla):")
print(df_filtered.head())

print("\nADIM 2: Veri Temizliği ve Ön İşleme Tamamlandı.")
```

```python
import pandas as pd
import numpy as np

print("ADIM 3: Film Öneri Sistemi Mantığı Oluşturuluyor...")

# ADIM 2'den df_filtered DataFrame'ini kullanıyoruz.
# df_filtered'ı global olarak tanımlayalım ki fonksiyonlar erişebilsin.
# Eğer bu kodu ayrı bir hücrede çalıştırıyorsanız, df_filtered'ın önceki hücrelerden geldiğinden emin olun.

# Kullanılabilir tüm türleri, yönetmenleri ve oyuncuları toplayalım (Gradio arayüzü için)
all_tags = sorted(list(set([tag for sublist in df_filtered['Tags_cleaned'] for tag in sublist])))
all_directors = sorted(list(set([director for sublist in df_filtered['Director_cleaned'] for director in sublist])))
all_stars = sorted(list(set([star for sublist in df_filtered['Stars_cleaned'] for star in sublist])))

# Kullanıcının seçtiği özelliklere göre film önerileri yapan ana fonksiyon
def get_movie_recommendations(selected_tags, selected_directors, selected_stars, min_imdb_rating=7.5, num_recommendations=10):

    # Filtreleme için temel DataFrame
    recommendations_df = df_filtered.copy()

    # 1. IMDb Puanına Göre Filtreleme
    recommendations_df = recommendations_df[recommendations_df['IMDb Rating'] >= min_imdb_rating]

    # 2. Türlere Göre Filtreleme
    if selected_tags:
        # Seçilen türlerden herhangi birini içeren filmleri bul
        # Her filmin etiket listesinde, seçilen etiketlerden en az biri olmalı
        recommendations_df = recommendations_df[
            recommendations_df['Tags_cleaned'].apply(lambda x: any(tag in x for tag in selected_tags))
        ]

    # 3. Yönetmenlere Göre Filtreleme
    if selected_directors:
        recommendations_df = recommendations_df[
            recommendations_df['Director_cleaned'].apply(lambda x: any(director in x for director in selected_directors))
        ]

    # 4. Oyunculara Göre Filtreleme
    if selected_stars:
        recommendations_df = recommendations_df[
            recommendations_df['Stars_cleaned'].apply(lambda x: any(star in x for star in selected_stars))
        ]

    # 5. Sonuçları IMDb Puanına (ve Votes_numeric'e) göre sıralama
    # Önce IMDb Rating'e, sonra Votes_numeric'e göre azalan sırada sırala
    recommendations_df = recommendations_df.sort_values(
        by=['IMDb Rating', 'Votes_numeric'],
        ascending=[False, False]
    ).reset_index(drop=True)

    # Sadece belirli sayıda öneri döndür
    top_recommendations = recommendations_df.head(num_recommendations)

    # Öneri sonuçlarını daha okunabilir bir formatta hazırla
    if top_recommendations.empty:
        return "Üzgünüz, seçtiğiniz kriterlere uygun film bulunamadı."
    else:
        # IMDb puanını ve oy sayısını formatlayalım
        top_recommendations['IMDb Rating'] = top_recommendations['IMDb Rating'].round(1).astype(str)
        top_recommendations['Votes'] = top_recommendations['Votes_numeric'].apply(lambda x: f"{int(x):,}")

        # Sadece 'Title', 'IMDb Rating', 'Director', 'Stars', 'Tags' sütunlarını gösterelim
        # Bu sütunları birleştirelim
        recommendation_list = []
        for idx, row in top_recommendations.iterrows():
            # Director ve Stars temizlenmiş listelerden stringe dönüştürülürken düzgün formatlansın
            directors_str = ", ".join([d.title() for d in row['Director_cleaned']]) # Her kelimenin ilk harfi büyük
            stars_str = ", ".join([s.title() for s in row['Stars_cleaned']])
            tags_str = ", ".join([t.title() for t in row['Tags_cleaned']])

            recommendation_list.append(
                f"**{row['Title']}** (IMDb: {row['IMDb Rating']} ⭐️, Oylar: {row['Votes']} 🗳️)"
                f"\n  Yönetmen: {directors_str}"
                f"\n  Oyuncular: {stars_str}"
                f"\n  Türler: {tags_str}"
                f"\n---"
            )
        return "\n".join(recommendation_list)

print("Örnek bir öneri sistemi testi yapılıyor...")
# Örnek test: Sadece "action" ve "drama" türünde filmler, IMDb puanı 8.0 ve üzeri
test_recs = get_movie_recommendations(
    selected_tags=['action', 'drama'],
    selected_directors=[],
    selected_stars=[],
    min_imdb_rating=8.0,
    num_recommendations=5
)
print("\nÖrnek Öneriler (Action & Drama, IMDb >= 8.0):")
print(test_recs)

print("\nADIM 3: Film Öneri Sistemi Mantığı Oluşturuldu.")
```

```python
import gradio as gr
import pandas as pd
import numpy as np
import os

if 'df_filtered' not in globals():
    print("HATA: df_filtered DataFrame'i bulunamadı.")
    print("Lütfen bu hücreyi çalıştırmadan önce ADIM 1, ADIM 2 ve ADIM 3 hücrelerini sırasıyla çalıştırdığınızdan emin olun.")
    exit()

all_tags = sorted(list(set([tag for sublist in df_filtered['Tags_cleaned'] for tag in sublist])))
all_directors = sorted(list(set([director for sublist in df_filtered['Director_cleaned'] for director in sublist])))
all_stars = sorted(list(set([star for sublist in df_filtered['Stars_cleaned'] for star in sublist])))

def get_movie_recommendations(selected_tags, selected_directors, selected_stars, min_imdb_rating_slider, num_recommendations_slider):

    print(f"\n--- Öneri İsteği ---")
    print(f"Seçilen Türler: {selected_tags}")
    print(f"Seçilen Yönetmenler: {selected_directors}")
    print(f"Seçilen Oyuncular: {selected_stars}")
    print(f"Minimum IMDb Puanı: {min_imdb_rating_slider}")
    print(f"Öneri Sayısı: {num_recommendations_slider}")
    print(f"Başlangıç DataFrame boyutu: {len(df_filtered)}")

    selected_tags = list(selected_tags) if selected_tags else []
    selected_directors = list(selected_directors) if selected_directors else []
    selected_stars = list(selected_stars) if selected_stars else []

    recommendations_df = df_filtered.copy()

    recommendations_df = recommendations_df[recommendations_df['IMDb Rating'] >= min_imdb_rating_slider]
    print(f"IMDb Puanı filtrelemesi sonrası: {len(recommendations_df)} film")

    if selected_tags:
        recommendations_df = recommendations_df[
            recommendations_df['Tags_cleaned'].apply(lambda x: any(tag in x for tag in selected_tags))
        ]
        print(f"Tür filtrelemesi sonrası: {len(recommendations_df)} film")

    if selected_directors:
        recommendations_df = recommendations_df[
            recommendations_df['Director_cleaned'].apply(lambda x: any(director in x for director in selected_directors))
        ]
        print(f"Yönetmen filtrelemesi sonrası: {len(recommendations_df)} film")

    if selected_stars:
        recommendations_df = recommendations_df[
            recommendations_df['Stars_cleaned'].apply(lambda x: any(star in x for star in selected_stars))
        ]
        print(f"Oyuncu filtrelemesi sonrası: {len(recommendations_df)} film")

    recommendations_df = recommendations_df.sort_values(
        by=['IMDb Rating', 'Votes_numeric'],
        ascending=[False, False]
    ).reset_index(drop=True)

    top_recommendations = recommendations_df.head(num_recommendations_slider)

    if top_recommendations.empty:
        print("Kriterlere uygun film bulunamadı.")
        return "Üzgünüz, seçtiğiniz kriterlere uygun film bulunamadı."
    else:
        print(f"Toplam {len(top_recommendations)} öneri bulundu.")
        recommendation_list = []
        for idx, row in top_recommendations.iterrows():
            directors_str = ", ".join([d.title() for d in row['Director_cleaned']])
            stars_str = ", ".join([s.title() for s in row['Stars_cleaned']])
            tags_str = ", ".join([t.title() for t in row['Tags_cleaned']])

            recommendation_list.append(
                f"**{row['Title']}** (IMDb: {row['IMDb Rating']:.1f} ⭐️, Oylar: {int(row['Votes_numeric']):,} 🗳️)"
                f"\n  Yönetmen: {directors_str if directors_str else 'Bilinmiyor'}"
                f"\n  Oyuncular: {stars_str if stars_str else 'Bilinmiyor'}"
                f"\n  Türler: {tags_str if tags_str else 'Bilinmiyor'}"
                f"\n---"
            )
        return "\n".join(recommendation_list)

    with gr.Blocks(theme=gr.themes.Soft()) as demo:
        gr.Markdown(
            """
            # 🎬 Film Öneri Sistemi
            Favori film özelliklerinizi seçin, yüksek IMDb puanına sahip filmleri keşfedin!
            """
        )

        with gr.Row():
            with gr.Column():
                gr.Markdown("### Film Özelliklerini Seçin:")

                tags_input = gr.CheckboxGroup(
                    label="Film Türleri",
                    choices=all_tags,
                    value=['action', 'drama'],
                    interactive=True
                )

                directors_input = gr.Dropdown(
                    label="Yönetmenler",
                    choices=all_directors,
                    multiselect=True,
                    allow_custom_value=False,
                    interactive=True
                )

                stars_input = gr.Dropdown(
                    label="Oyuncular",
                    choices=all_stars,
                    multiselect=True,
                    allow_custom_value=False,
                    interactive=True
                )

                min_imdb_rating_slider = gr.Slider(
                    minimum=df_filtered['IMDb Rating'].min(),
                    maximum=df_filtered['IMDb Rating'].max(),
                    step=0.1,
                    value=7.6,
                    label="Minimum IMDb Puanı"
                )

                num_recommendations_slider = gr.Slider(
                    minimum=1,
                    maximum=20,
                    step=1,
                    value=10,
                    label="Öneri Sayısı"
                )

                recommend_btn = gr.Button("🚀 Film Önerilerini Getir", variant="primary", size="lg")

            with gr.Column():
                gr.Markdown("### Önerilen Filmler:")
                output_text = gr.Markdown(label="Önerileriniz burada listelenecektir.", value="Henüz bir öneri yapılmadı. Özellikleri seçip butona tıklayın!")

        gr.Examples(
            examples=[
                [['action'], [], [], 7.6, 5],
                [['comedy', 'drama'], [], [], 7.8, 3],
                [[], ['christopher nolan'], [], 8.0, 5],
                [[], [], ['leonardo dicaprio'], 7.8, 3],
            ],
            inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider],
            outputs=output_text,
            fn=get_movie_recommendations,
            label="Örnek Önerileri Deneyin"
        )

        gr.Markdown(
            """
            ---
            ### ℹ️ Nasıl Kullanılır?
            1.  **Film Türleri, Yönetmenler ve Oyuncular** bölümlerinden istediğiniz filtreleri seçin (birden fazla seçim yapabilirsiniz).
            2.  **Minimum IMDb Puanı** ve **Öneri Sayısı** çubuklarını ayarlayın.
            3.  **"Film Önerilerini Getir"** butonuna tıklayın.
            4.  Öneriler sağ panelde görünecektir!
            """
        )

    demo.launch(share=True)
```

```python
# Kodu çalıştırmadan önce:
# 1. Colab'da yeni bir not defteri açın.
# 2. Çalışma Zamanı -> Çalışma zamanı türünü değiştir -> Donanım hızlandırıcıyı "None" olarak ayarlayın.
# 3. 'imdb-top-rated-movies-user-rated.csv' dosyasını sol paneldeki dosya simgesini kullanarak Colab'a yükleyin.

import pandas as pd
import numpy as np
import gradio as gr
import os

print("--- Film Öneri Sistemi Başlatılıyor ---")

# --- ADIM 1: Veri Seti Yükleniyor ve Keşfediliyor ---
print("ADIM 1: Veri Seti Yükleniyor ve Keşfediliyor...")

csv_file_name = "imdb-top-rated-movies-user-rated.csv" # Yüklediğiniz CSV dosyasının adı
file_path = os.path.join("/content/", csv_file_name)

if not os.path.exists(file_path):
    print(f"HATA: '{csv_file_name}' dosyası '{file_path}' yolunda bulunamadı.")
    print("Lütfen CSV dosyasını Colab'a doğru şekilde yüklediğinizden ve 'csv_file_name' değişkenini doğru ayarladığınızdan emin olun.")
    exit()

df = pd.read_csv(file_path)

print(f"'{csv_file_name}' başarıyla yüklendi. Toplam {len(df)} film bulundu.")
print("\nVeri Setinin İlk 5 Satırı:")
print(df.head())
print("\nADIM 1: Veri Seti Keşfi Tamamlandı.")

# --- ADIM 2: Veri Temizliği ve Ön İşleme ---
print("\nADIM 2: Veri Temizliği ve Ön İşleme Başlıyor (Türler Basitleştiriliyor)...")

df_filtered = df[['Title', 'IMDb Rating', 'Tags', 'Director', 'Stars', 'Votes']].copy()

df_filtered['Stars'].fillna('', inplace=True)

# Türleri Basitleştirme için eşleme sözlüğü
genre_mapping = {
    'action': ['action', 'action epic', 'gun fu', 'one-person army action', 'car action', 'kung fu', 'martial arts'],
    'adventure': ['adventure', 'adventure epic', 'desert adventure', 'animal adventure', 'space adventure', 'swashbuckler'],
    'comedy': ['comedy', 'romantic comedy', 'buddy comedy', 'sitcom', 'black comedy', 'satire', 'spoof', 'parody', 'slapstick', 'screwball comedy', 'dark comedy', 'body swap comedy'],
    'drama': ['drama', 'period drama', 'cop drama', 'legal drama', 'medical drama', 'teen drama', 'psychological drama', 'melodrama', 'historical drama', 'biography', 'romantic drama', 'showbiz drama', 'tragedy'],
    'thriller': ['thriller', 'crime thriller', 'spy thriller', 'psychological thriller', 'mystery thriller', 'political thriller', 'conspiracy thriller', 'erotic thriller', 'cyber thriller', 'suspense'],
    'sci-fi': ['sci-fi', 'space sci-fi', 'dystopian sci-fi', 'cyberpunk', 'alien invasion', 'mutant', 'robot', 'post-apocalyptic', 'time travel'],
    'fantasy': ['fantasy', 'dark fantasy', 'sword & sorcery', 'fairy tale', 'epic fantasy'],
    'horror': ['horror', 'slasher', 'supernatural horror', 'body horror', 'zombie', 'monster', 'vampire', 'werewolf', 'ghost'],
    'mystery': ['mystery', 'suspense mystery', 'cozy mystery', 'whodunnit', 'detective', 'police procedural'],
    'crime': ['crime', 'gangster', 'heist', 'mob', 'true crime'],
    'romance': ['romance', 'romantic comedy', 'romantic drama'],
    'animation': ['animation', 'adult animation', 'anime', 'computer animation', 'drawn animation', 'stop-motion animation'],
    'family': ['family', 'kids'],
    'western': ['western', 'classic western', 'neo-western'],
    'war': ['war', 'war drama'],
    'history': ['history', 'historical drama', 'biography'],
    'music': ['music', 'musical', 'classic musical', 'concert'],
    'documentary': ['documentary', 'docudrama', 'mockumentary']
}

reverse_genre_map = {}
for main_genre, sub_genres in genre_mapping.items():
    for sub_genre in sub_genres:
        reverse_genre_map[sub_genre] = main_genre

def map_to_main_genres(tag_list):
    main_genres = set()
    for tag in tag_list:
        main_genres.add(reverse_genre_map.get(tag, tag))
    return list(main_genres)

def clean_and_split(text_series):
    if isinstance(text_series, float) and pd.isna(text_series): # NaN kontrolü
        return []

    # ensure it's a string before processing
    item = str(text_series)
    item = item.replace('"', '').replace("'", '').strip()
    split_items = [s.strip().lower() for s in item.split(',') if s.strip()]
    return split_items


df_filtered['Tags_cleaned_raw'] = df_filtered['Tags'].apply(clean_and_split)
df_filtered['Director_cleaned'] = df_filtered['Director'].apply(clean_and_split)
df_filtered['Stars_cleaned'] = df_filtered['Stars'].apply(clean_and_split)

df_filtered['Tags_cleaned'] = df_filtered['Tags_cleaned_raw'].apply(map_to_main_genres)


def convert_votes_to_numeric(votes_str):
    if isinstance(votes_str, str):
        votes_str = votes_str.replace(",", "") # Virgülleri temizle
        if 'K' in votes_str:
            return float(votes_str.replace('K', '')) * 1000
        elif 'M' in votes_str:
            return float(votes_str.replace('M', '')) * 1_000_000
    return float(votes_str)

df_filtered['Votes_numeric'] = df_filtered['Votes'].apply(convert_votes_to_numeric)
df_filtered.drop('Votes', axis=1, inplace=True)

print("\nTemizlenmiş Veri Setinin İlk 5 Satırı (Yeni Sütunlarla):")
print(df_filtered.head())
print("\nBasitleştirilmiş Türlerin Örnekleri:")
print(df_filtered['Tags_cleaned'].head())
print("\nADIM 2: Veri Temizliği ve Ön İşleme Tamamlandı.")

# --- ADIM 3: Film Öneri Sistemi Mantığını Oluşturma ---
print("\nADIM 3: Film Öneri Sistemi Mantığı Oluşturuluyor...")

all_tags = sorted(list(set([tag for sublist in df_filtered['Tags_cleaned'] for tag in sublist])))
all_directors = sorted(list(set([director for sublist in df_filtered['Director_cleaned'] for director in sublist])))
all_stars = sorted(list(set([star for sublist in df_filtered['Stars_cleaned'] for star in sublist])))

def get_movie_recommendations(selected_tags, selected_directors, selected_stars, min_imdb_rating_slider, num_recommendations_slider):

    print(f"\n--- Öneri İsteği ---")
    print(f"Seçilen Türler: {selected_tags}")
    print(f"Seçilen Yönetmenler: {selected_directors}")
    print(f"Seçilen Oyuncular: {selected_stars}")
    print(f"Minimum IMDb Puanı: {min_imdb_rating_slider}")
    print(f"Öneri Sayısı: {num_recommendations_slider}")
    print(f"Başlangıç DataFrame boyutu: {len(df_filtered)}")

    selected_tags = list(selected_tags) if selected_tags else []
    selected_directors = list(selected_directors) if selected_directors else []
    selected_stars = list(selected_stars) if selected_stars else []

    recommendations_df = df_filtered.copy()

    recommendations_df = recommendations_df[recommendations_df['IMDb Rating'] >= min_imdb_rating_slider]
    print(f"IMDb Puanı filtrelemesi sonrası: {len(recommendations_df)} film")

    if selected_tags:
        recommendations_df = recommendations_df[
            recommendations_df['Tags_cleaned'].apply(lambda x: any(tag in x for tag in selected_tags))
        ]
        print(f"Tür filtrelemesi sonrası: {len(recommendations_df)} film")

    if selected_directors:
        recommendations_df = recommendations_df[
            recommendations_df['Director_cleaned'].apply(lambda x: any(director in x for director in selected_directors))
        ]
        print(f"Yönetmen filtrelemesi sonrası: {len(recommendations_df)} film")

    if selected_stars:
        recommendations_df = recommendations_df[
            recommendations_df['Stars_cleaned'].apply(lambda x: any(star in x for star in selected_stars))
        ]
        print(f"Oyuncu filtrelemesi sonrası: {len(recommendations_df)} film")

    recommendations_df = recommendations_df.sort_values(
        by=['IMDb Rating', 'Votes_numeric'],
        ascending=[False, False]
    ).reset_index(drop=True)

    top_recommendations = recommendations_df.head(num_recommendations_slider)

    if top_recommendations.empty:
        print("Kriterlere uygun film bulunamadı.")
        return "Üzgünüz, seçtiğiniz kriterlere uygun film bulunamadı."
    else:
        print(f"Toplam {len(top_recommendations)} öneri bulundu.")
        recommendation_list = []
        for idx, row in top_recommendations.iterrows():
            directors_str = ", ".join([d.title() for d in row['Director_cleaned']])
            stars_str = ", ".join([s.title() for s in row['Stars_cleaned']])
            tags_str = ", ".join([t.title() for t in row['Tags_cleaned']])

            recommendation_list.append(
                f"**{row['Title']}** (IMDb: {row['IMDb Rating']:.1f} ⭐️, Oylar: {int(row['Votes_numeric']):,} 🗳️)"
                f"\n  Yönetmen: {directors_str if directors_str else 'Bilinmiyor'}"
                f"\n  Oyuncular: {stars_str if stars_str else 'Bilinmiyor'}"
                f"\n  Türler: {tags_str if tags_str else 'Bilinmiyor'}"
                f"\n---"
            )
        return "\n".join(recommendation_list)

print("\nADIM 3: Film Öneri Sistemi Mantığı Oluşturuldu.")

# --- ADIM 4: Gradio Web Arayüzü Oluşturma ---
print("\nADIM 4: Gradio Web Arayüzü Oluşturuluyor...")

with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown(
        """
        # 🎬 Film Öneri Sistemi
        Favori film özelliklerinizi seçin, yüksek IMDb puanına sahip filmleri keşfedin!
        """
    )

    with gr.Row():
        with gr.Column():
            gr.Markdown("### Film Özelliklerini Seçin:")

            tags_input = gr.CheckboxGroup(
                label="Film Türleri",
                choices=all_tags,
                value=['action', 'drama'],
                interactive=True
            )

            directors_input = gr.Dropdown(
                label="Yönetmenler",
                choices=all_directors,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            stars_input = gr.Dropdown(
                label="Oyuncular",
                choices=all_stars,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            min_imdb_rating_slider = gr.Slider(
                minimum=df_filtered['IMDb Rating'].min(),
                maximum=df_filtered['IMDb Rating'].max(),
                step=0.1,
                value=7.6,
                label="Minimum IMDb Puanı"
            )

            num_recommendations_slider = gr.Slider(
                minimum=1,
                maximum=20,
                step=1,
                value=10,
                label="Öneri Sayısı"
            )

            recommend_btn = gr.Button("🚀 Film Önerilerini Getir", variant="primary", size="lg")

        with gr.Column():
            gr.Markdown("### Önerilen Filmler:")
            output_text = gr.Markdown(label="Önerileriniz burada listelenecektir.", value="Henüz bir öneri yapılmadı. Özellikleri seçip butona tıklayın!")

    gr.Examples(
        examples=[
            [['action'], [], [], 7.6, 5],
            [['comedy', 'drama'], [], [], 7.8, 3],
            [[], ['christopher nolan'], [], 8.0, 5],
            [[], [], ['leonardo dicaprio'], 7.8, 3],
        ],
        inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider],
        outputs=output_text,
        fn=get_movie_recommendations,
        label="Örnek Önerileri Deneyin"
    )

    gr.Markdown(
        """
        ---
        ### ℹ️ Nasıl Kullanılır?
        1.  **Film Türleri, Yönetmenler ve Oyuncular** bölümlerinden istediğiniz filtreleri seçin (birden fazla seçim yapabilirsiniz).
        2.  **Minimum IMDb Puanı** ve **Öneri Sayısı** çubuklarını ayarlayın.
        3.  **"Film Önerilerini Getir"** butonuna tıklayın.
        4.  Öneriler sağ panelde görünecektir!
        """
    )

demo.launch(share=True)
print("\nADIM 4: Gradio Web Arayüzü Başlatıldı.")
```
