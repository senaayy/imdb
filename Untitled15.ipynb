```python
import pandas as pd
import numpy as np
import gradio as gr
import os

print("ADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor...")

# Y√ºklediƒüiniz CSV dosyasƒ±nƒ±n adƒ±nƒ± buraya yazƒ±n
# Ekran g√∂r√ºnt√ºn√ºzden dosya adƒ±nƒ±n 'imdb-top-rated-movies-user-rated.csv' olduƒüunu g√∂r√ºyorum.
csv_file_name = "imdb-top-rated-movies-user-rated.csv"

# Colab'daki varsayƒ±lan y√ºkleme konumu /content/ olduƒüundan
file_path = os.path.join("/content/", csv_file_name)

if not os.path.exists(file_path):
    print(f"HATA: '{csv_file_name}' dosyasƒ± '{file_path}' yolunda bulunamadƒ±.")
    print("L√ºtfen CSV dosyasƒ±nƒ± Colab'a doƒüru ≈üekilde y√ºklediƒüinizden ve 'csv_file_name' deƒüi≈ükenini doƒüru ayarladƒ±ƒüƒ±nƒ±zdan emin olun.")
    exit()

df = pd.read_csv(file_path)

print(f"'{csv_file_name}' ba≈üarƒ±yla y√ºklendi. Toplam {len(df)} film bulundu.")
print("\nVeri Setinin ƒ∞lk 5 Satƒ±rƒ±:")
print(df.head())

print("\nVeri Setindeki S√ºtunlar ve Veri Tipleri:")
print(df.info())

print("\nIMDb Puanƒ± Daƒüƒ±lƒ±mƒ±:")
# Ekran g√∂r√ºnt√ºn√ºzdeki s√ºtun ba≈ülƒ±ƒüƒ± 'IMDb Rating' (Bo≈üluklu ve b√ºy√ºk harfli) olduƒüu i√ßin onu kullanƒ±yoruz.
print(df['IMDb Rating'].describe())

print("\nADIM 1: Veri Seti Ke≈üfi Tamamlandƒ±.")
```

```python
import pandas as pd
import numpy as np

print("ADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Ba≈ülƒ±yor...")

# ADIM 1'den df DataFrame'ini kullanƒ±yoruz

# 1. ƒ∞htiyacƒ±mƒ±z Olan S√ºtunlarƒ± Se√ßme
# √ñneri sistemi i√ßin kullanacaƒüƒ±mƒ±z s√ºtunlar:
# Title (Ba≈ülƒ±k), IMDb Rating (IMDb Puanƒ±), Tags (T√ºrler), Director (Y√∂netmen), Stars (Oyuncular), Votes (Oy Sayƒ±sƒ±)
df_filtered = df[['Title', 'IMDb Rating', 'Tags', 'Director', 'Stars', 'Votes']].copy()

# 2. Eksik Verileri ƒ∞≈üleme (NaN deƒüerler)
# IMDb Rating, Tags, Director s√ºtunlarƒ±nda eksik veri olmamalƒ± (df.info() √ßƒ±ktƒ±sƒ±na g√∂re yok)
# Stars s√ºtununda 1 eksik deƒüer vardƒ±. Bu t√ºr k√º√ß√ºk eksiklikleri bo≈ü string ile doldurabiliriz.
df_filtered['Stars'].fillna('', inplace=True)

# Eksik deƒüerleri doldurduktan sonra tekrar kontrol edelim
print("\nEksik deƒüer kontrol√º (fillna sonrasƒ±):")
print(df_filtered.isnull().sum())

# 3. 'Tags', 'Director', 'Stars' S√ºtunlarƒ±nƒ± Standartla≈ütƒ±rma ve Temizleme
# Bu s√ºtunlar genellikle virg√ºlle ayrƒ±lmƒ±≈ü birden fazla deƒüeri i√ßerir (√∂rn. "Action, Drama")
# Bazƒ±larƒ± tƒ±rnak i√ßinde olabilir, onlarƒ± da temizleyelim.

def clean_and_split(text_series):
    # NaN ise bo≈ü liste d√∂nd√ºr
    if text_series.isnull().any():
        text_series = text_series.fillna('')

    # Her bir √∂ƒüeyi temizle: tƒ±rnaklarƒ± kaldƒ±r, bo≈üluklarƒ± d√ºzelt, k√º√ß√ºk harfe √ßevir
    cleaned_list = []
    for item in text_series.astype(str): # Hata vermemesi i√ßin stringe √ßevir
        item = item.replace('"', '').replace("'", '').strip() # Tƒ±rnaklarƒ± ve gereksiz bo≈üluklarƒ± kaldƒ±r
        split_items = [s.strip().lower() for s in item.split(',') if s.strip()] # Virg√ºlle ayƒ±r, bo≈üluklarƒ± d√ºzelt, k√º√ß√ºk harfe √ßevir
        cleaned_list.append(split_items)
    return cleaned_list

df_filtered['Tags_cleaned'] = clean_and_split(df_filtered['Tags'])
df_filtered['Director_cleaned'] = clean_and_split(df_filtered['Director'])
df_filtered['Stars_cleaned'] = clean_and_split(df_filtered['Stars'])

# 4. 'Votes' S√ºtununu Sayƒ±sal Deƒüere √áevirme
# "927K" gibi deƒüerler var. 'K' varsa 1000 ile √ßarpacaƒüƒ±z.
def convert_votes_to_numeric(votes_str):
    if isinstance(votes_str, str):
        if 'K' in votes_str:
            return float(votes_str.replace('K', '')) * 1000
        elif 'M' in votes_str: # Milyon i√ßin de eklendi
            return float(votes_str.replace('M', '')) * 1_000_000
    return float(votes_str) # Zaten sayƒ±sal ise veya ba≈üka bir formatta ise direkt √ßevir

df_filtered['Votes_numeric'] = df_filtered['Votes'].apply(convert_votes_to_numeric)

# Gereksiz Votes s√ºtununu kaldƒ±r
df_filtered.drop('Votes', axis=1, inplace=True)


print("\nTemizlenmi≈ü Veri Setinin ƒ∞lk 5 Satƒ±rƒ± (Yeni S√ºtunlarla):")
print(df_filtered.head())

print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Tamamlandƒ±.")
```

```python
import pandas as pd
import numpy as np

print("ADIM 3: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuluyor...")

# ADIM 2'den df_filtered DataFrame'ini kullanƒ±yoruz.
# df_filtered'ƒ± global olarak tanƒ±mlayalƒ±m ki fonksiyonlar eri≈üebilsin.
# Eƒüer bu kodu ayrƒ± bir h√ºcrede √ßalƒ±≈ütƒ±rƒ±yorsanƒ±z, df_filtered'ƒ±n √∂nceki h√ºcrelerden geldiƒüinden emin olun.

# Kullanƒ±labilir t√ºm t√ºrleri, y√∂netmenleri ve oyuncularƒ± toplayalƒ±m (Gradio aray√ºz√º i√ßin)
all_tags = sorted(list(set([tag for sublist in df_filtered['Tags_cleaned'] for tag in sublist])))
all_directors = sorted(list(set([director for sublist in df_filtered['Director_cleaned'] for director in sublist])))
all_stars = sorted(list(set([star for sublist in df_filtered['Stars_cleaned'] for star in sublist])))

# Kullanƒ±cƒ±nƒ±n se√ßtiƒüi √∂zelliklere g√∂re film √∂nerileri yapan ana fonksiyon
def get_movie_recommendations(selected_tags, selected_directors, selected_stars, min_imdb_rating=7.5, num_recommendations=10):

    # Filtreleme i√ßin temel DataFrame
    recommendations_df = df_filtered.copy()

    # 1. IMDb Puanƒ±na G√∂re Filtreleme
    recommendations_df = recommendations_df[recommendations_df['IMDb Rating'] >= min_imdb_rating]

    # 2. T√ºrlere G√∂re Filtreleme
    if selected_tags:
        # Se√ßilen t√ºrlerden herhangi birini i√ßeren filmleri bul
        # Her filmin etiket listesinde, se√ßilen etiketlerden en az biri olmalƒ±
        recommendations_df = recommendations_df[
            recommendations_df['Tags_cleaned'].apply(lambda x: any(tag in x for tag in selected_tags))
        ]

    # 3. Y√∂netmenlere G√∂re Filtreleme
    if selected_directors:
        recommendations_df = recommendations_df[
            recommendations_df['Director_cleaned'].apply(lambda x: any(director in x for director in selected_directors))
        ]

    # 4. Oyunculara G√∂re Filtreleme
    if selected_stars:
        recommendations_df = recommendations_df[
            recommendations_df['Stars_cleaned'].apply(lambda x: any(star in x for star in selected_stars))
        ]

    # 5. Sonu√ßlarƒ± IMDb Puanƒ±na (ve Votes_numeric'e) g√∂re sƒ±ralama
    # √ñnce IMDb Rating'e, sonra Votes_numeric'e g√∂re azalan sƒ±rada sƒ±rala
    recommendations_df = recommendations_df.sort_values(
        by=['IMDb Rating', 'Votes_numeric'],
        ascending=[False, False]
    ).reset_index(drop=True)

    # Sadece belirli sayƒ±da √∂neri d√∂nd√ºr
    top_recommendations = recommendations_df.head(num_recommendations)

    # √ñneri sonu√ßlarƒ±nƒ± daha okunabilir bir formatta hazƒ±rla
    if top_recommendations.empty:
        return "√úzg√ºn√ºz, se√ßtiƒüiniz kriterlere uygun film bulunamadƒ±."
    else:
        # IMDb puanƒ±nƒ± ve oy sayƒ±sƒ±nƒ± formatlayalƒ±m
        top_recommendations['IMDb Rating'] = top_recommendations['IMDb Rating'].round(1).astype(str)
        top_recommendations['Votes'] = top_recommendations['Votes_numeric'].apply(lambda x: f"{int(x):,}")

        # Sadece 'Title', 'IMDb Rating', 'Director', 'Stars', 'Tags' s√ºtunlarƒ±nƒ± g√∂sterelim
        # Bu s√ºtunlarƒ± birle≈ütirelim
        recommendation_list = []
        for idx, row in top_recommendations.iterrows():
            # Director ve Stars temizlenmi≈ü listelerden stringe d√∂n√º≈üt√ºr√ºl√ºrken d√ºzg√ºn formatlansƒ±n
            directors_str = ", ".join([d.title() for d in row['Director_cleaned']]) # Her kelimenin ilk harfi b√ºy√ºk
            stars_str = ", ".join([s.title() for s in row['Stars_cleaned']])
            tags_str = ", ".join([t.title() for t in row['Tags_cleaned']])

            recommendation_list.append(
                f"**{row['Title']}** (IMDb: {row['IMDb Rating']} ‚≠êÔ∏è, Oylar: {row['Votes']} üó≥Ô∏è)"
                f"\n  Y√∂netmen: {directors_str}"
                f"\n  Oyuncular: {stars_str}"
                f"\n  T√ºrler: {tags_str}"
                f"\n---"
            )
        return "\n".join(recommendation_list)

print("√ñrnek bir √∂neri sistemi testi yapƒ±lƒ±yor...")
# √ñrnek test: Sadece "action" ve "drama" t√ºr√ºnde filmler, IMDb puanƒ± 8.0 ve √ºzeri
test_recs = get_movie_recommendations(
    selected_tags=['action', 'drama'],
    selected_directors=[],
    selected_stars=[],
    min_imdb_rating=8.0,
    num_recommendations=5
)
print("\n√ñrnek √ñneriler (Action & Drama, IMDb >= 8.0):")
print(test_recs)

print("\nADIM 3: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuldu.")
```

```python
# SADECE BU √áER√áEVE ƒ∞√áƒ∞NDEKƒ∞ KODU KOPYALAYIN VE YAPI≈ûTIRIN!

import gradio as gr
import pandas as pd
import numpy as np

print("ADIM 4: Gradio Web Aray√ºz√º Olu≈üturuluyor (Hata Ayƒ±klama Aktif)...")

# √ñNEMLƒ∞: Eƒüer Colab'ƒ± kapatƒ±p a√ßtƒ±ysanƒ±z veya √ßalƒ±≈üma zamanƒ±nƒ± yeniden ba≈ülattƒ±ysanƒ±z,
# bu h√ºcreyi √ßalƒ±≈ütƒ±rmadan √∂nce ADIM 1, ADIM 2 ve ADIM 3 h√ºcrelerini tekrar √áALI≈ûTIRDIƒûINIZDAN
# emin olun. Aksi takdirde df, df_filtered vb. deƒüi≈ükenler tanƒ±msƒ±z kalƒ±r.

# df_filtered'ƒ±n tanƒ±mlƒ± olduƒüunu kontrol edelim, deƒüilse hata verelim
if 'df_filtered' not in globals():
    print("HATA: df_filtered DataFrame'i bulunamadƒ±.")
    print("L√ºtfen bu h√ºcreyi √ßalƒ±≈ütƒ±rmadan √∂nce ADIM 1, ADIM 2 ve ADIM 3 h√ºcrelerini sƒ±rasƒ±yla √ßalƒ±≈ütƒ±rdƒ±ƒüƒ±nƒ±zdan emin olun.")
    exit() # DataFrame yoksa programƒ± durdur

# Kullanƒ±labilir t√ºm t√ºrleri, y√∂netmenleri ve oyuncularƒ± toplayalƒ±m (Gradio aray√ºz√º i√ßin)
# Bu listeler ADIM 2 ve ADIM 3'teki df_filtered'dan olu≈üturulur.
all_tags = sorted(list(set([tag for sublist in df_filtered['Tags_cleaned'] for tag in sublist])))
all_directors = sorted(list(set([director for sublist in df_filtered['Director_cleaned'] for director in sublist])))
all_stars = sorted(list(set([star for sublist in df_filtered['Stars_cleaned'] for star in sublist])))

# Kullanƒ±cƒ±nƒ±n se√ßtiƒüi √∂zelliklere g√∂re film √∂nerileri yapan ana fonksiyon
def get_movie_recommendations(selected_tags, selected_directors, selected_stars, min_imdb_rating_slider, num_recommendations_slider):

    # Hata ayƒ±klama √ßƒ±ktƒ±larƒ±
    print(f"\n--- √ñneri ƒ∞steƒüi ---")
    print(f"Se√ßilen T√ºrler: {selected_tags}")
    print(f"Se√ßilen Y√∂netmenler: {selected_directors}")
    print(f"Se√ßilen Oyuncular: {selected_stars}")
    print(f"Minimum IMDb Puanƒ±: {min_imdb_rating_slider}")
    print(f"√ñneri Sayƒ±sƒ±: {num_recommendations_slider}")
    print(f"Ba≈ülangƒ±√ß DataFrame boyutu: {len(df_filtered)}")

    # Gradio'dan gelen se√ßili deƒüerler tuple olarak gelebilir, listeye √ßevirelim
    selected_tags = list(selected_tags) if selected_tags else []
    selected_directors = list(selected_directors) if selected_directors else []
    selected_stars = list(selected_stars) if selected_stars else []

    recommendations_df = df_filtered.copy()

    # 1. IMDb Puanƒ±na G√∂re Filtreleme (Slider deƒüeri kullanƒ±lƒ±yor)
    recommendations_df = recommendations_df[recommendations_df['IMDb Rating'] >= min_imdb_rating_slider]
    print(f"IMDb Puanƒ± filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # 2. T√ºrlere G√∂re Filtreleme
    if selected_tags:
        recommendations_df = recommendations_df[
            recommendations_df['Tags_cleaned'].apply(lambda x: any(tag in x for tag in selected_tags))
        ]
        print(f"T√ºr filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # 3. Y√∂netmenlere G√∂re Filtreleme
    if selected_directors:
        recommendations_df = recommendations_df[
            recommendations_df['Director_cleaned'].apply(lambda x: any(director in x for director in selected_directors))
        ]
        print(f"Y√∂netmen filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # 4. Oyunculara G√∂re Filtreleme
    if selected_stars:
        recommendations_df = recommendations_df[
            recommendations_df['Stars_cleaned'].apply(lambda x: any(star in x for star in selected_stars))
        ]
        print(f"Oyuncu filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    # 5. Sonu√ßlarƒ± IMDb Puanƒ±na (ve Votes_numeric'e) g√∂re sƒ±ralama
    recommendations_df = recommendations_df.sort_values(
        by=['IMDb Rating', 'Votes_numeric'],
        ascending=[False, False]
    ).reset_index(drop=True)

    # Sadece belirli sayƒ±da √∂neri d√∂nd√ºr (Slider deƒüeri kullanƒ±lƒ±yor)
    top_recommendations = recommendations_df.head(num_recommendations_slider)

    if top_recommendations.empty:
        print("Kriterlere uygun film bulunamadƒ±.")
        return "√úzg√ºn√ºz, se√ßtiƒüiniz kriterlere uygun film bulunamadƒ±."
    else:
        print(f"Toplam {len(top_recommendations)} √∂neri bulundu.")
        recommendation_list = []
        for idx, row in top_recommendations.iterrows():
            directors_str = ", ".join([d.title() for d in row['Director_cleaned']])
            stars_str = ", ".join([s.title() for s in row['Stars_cleaned']])
            tags_str = ", ".join([t.title() for t in row['Tags_cleaned']])

            recommendation_list.append(
                f"**{row['Title']}** (IMDb: {row['IMDb Rating']:.1f} ‚≠êÔ∏è, Oylar: {int(row['Votes_numeric']):,} üó≥Ô∏è)"
                f"\n  Y√∂netmen: {directors_str if directors_str else 'Bilinmiyor'}"
                f"\n  Oyuncular: {stars_str if stars_str else 'Bilinmiyor'}"
                f"\n  T√ºrler: {tags_str if tags_str else 'Bilinmiyor'}"
                f"\n---"
            )
        return "\n".join(recommendation_list)

    # Gradio Aray√ºz Tanƒ±mƒ±
    with gr.Blocks(theme=gr.themes.Soft()) as demo:
        gr.Markdown(
            """
            # üé¨ Film √ñneri Sistemi
            Favori film √∂zelliklerinizi se√ßin, y√ºksek IMDb puanƒ±na sahip filmleri ke≈üfedin!
            """
        )

        with gr.Row():
            with gr.Column():
                gr.Markdown("### Film √ñzelliklerini Se√ßin:")

                # T√ºr se√ßimi (√áoklu se√ßim)
                tags_input = gr.CheckboxGroup(
                    label="Film T√ºrleri",
                    choices=all_tags, # all_tags listesi buradan geliyor
                    value=['action', 'drama'], # Varsayƒ±lan deƒüerler
                    interactive=True
                )

                # Y√∂netmen se√ßimi (√áoklu se√ßim)
                directors_input = gr.Dropdown(
                    label="Y√∂netmenler",
                    choices=all_directors, # all_directors listesi buradan geliyor
                    multiselect=True,
                    allow_custom_value=False,
                    interactive=True
                )

                # Oyuncu se√ßimi (√áoklu se√ßim)
                stars_input = gr.Dropdown(
                    label="Oyuncular",
                    choices=all_stars, # all_stars listesi buradan geliyor
                    multiselect=True,
                    allow_custom_value=False,
                    interactive=True
                )

                # Minimum IMDb puanƒ± slider'ƒ±
                min_imdb_rating_slider = gr.Slider(
                    minimum=df_filtered['IMDb Rating'].min(),
                    maximum=df_filtered['IMDb Rating'].max(),
                    step=0.1,
                    value=7.6, # Ba≈ülangƒ±√ßta en d√º≈ü√ºk puanƒ± varsayƒ±lan yapalƒ±m, b√∂ylece hep sonu√ß gelir
                    label="Minimum IMDb Puanƒ±"
                )

                # √ñneri sayƒ±sƒ± slider'ƒ±
                num_recommendations_slider = gr.Slider(
                    minimum=1,
                    maximum=20,
                    step=1,
                    value=10, # Varsayƒ±lan √∂neri sayƒ±sƒ±
                    label="√ñneri Sayƒ±sƒ±"
                )

                recommend_btn = gr.Button("üöÄ Film √ñnerilerini Getir", variant="primary", size="lg")

            with gr.Column():
                gr.Markdown("### √ñnerilen Filmler:")
                output_text = gr.Markdown(label="√ñnerileriniz burada listelenecektir.", value="Hen√ºz bir √∂neri yapƒ±lmadƒ±. √ñzellikleri se√ßip butona tƒ±klayƒ±n!")

        # √ñrnekler ekleyelim (Default deƒüerleri biraz daha esnek yapƒ±ldƒ±)
        gr.Examples(
            examples=[
                [['action'], [], [], 7.6, 5], # Sadece Aksiyon, en d√º≈ü√ºk puanla
                [['comedy', 'drama'], [], [], 7.8, 3], # Komedi/Drama, biraz daha y√ºksek puan
                [[], ['christopher nolan'], [], 8.0, 5], # Sadece Christopher Nolan filmleri
                [[], [], ['leonardo dicaprio'], 7.8, 3], # Sadece Leonardo DiCaprio filmleri
            ],
            inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider],
            outputs=output_text,
            fn=get_movie_recommendations,
            label="√ñrnek √ñnerileri Deneyin"
        )

        gr.Markdown(
            """
            ---
            ### ‚ÑπÔ∏è Nasƒ±l Kullanƒ±lƒ±r?
            1.  **Film T√ºrleri, Y√∂netmenler ve Oyuncular** b√∂l√ºmlerinden istediƒüiniz filtreleri se√ßin (birden fazla se√ßim yapabilirsiniz).
            2.  **Minimum IMDb Puanƒ±** ve **√ñneri Sayƒ±sƒ±** √ßubuklarƒ±nƒ± ayarlayƒ±n.
            3.  **"Film √ñnerilerini Getir"** butonuna tƒ±klayƒ±n.
            4.  √ñneriler saƒü panelde g√∂r√ºnecektir!
            """
        )

    # Uygulamayƒ± ba≈ülat
    demo.launch(share=True) # share=True ile herkesin eri≈üebileceƒüi bir link olu≈üturur

# SADECE BU √áER√áEVE ƒ∞√áƒ∞NDEKƒ∞ KODU KOPYALAYIN VE YAPI≈ûTIRIN!
```

```python
# Kodu √ßalƒ±≈ütƒ±rmadan √∂nce:
# 1. Colab'da yeni bir not defteri a√ßƒ±n.
# 2. √áalƒ±≈üma Zamanƒ± -> √áalƒ±≈üma zamanƒ± t√ºr√ºn√º deƒüi≈ütir -> Donanƒ±m hƒ±zlandƒ±rƒ±cƒ±yƒ± "None" olarak ayarlayƒ±n.
# 3. 'imdb-top-rated-movies-user-rated.csv' dosyasƒ±nƒ± sol paneldeki dosya simgesini kullanarak Colab'a y√ºkleyin.

import pandas as pd
import numpy as np
import gradio as gr
import os

print("--- Film √ñneri Sistemi Ba≈ülatƒ±lƒ±yor ---")

# --- ADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor ---
print("ADIM 1: Veri Seti Y√ºkleniyor ve Ke≈üfediliyor...")

csv_file_name = "imdb-top-rated-movies-user-rated.csv" # Y√ºklediƒüiniz CSV dosyasƒ±nƒ±n adƒ±
file_path = os.path.join("/content/", csv_file_name)

if not os.path.exists(file_path):
    print(f"HATA: '{csv_file_name}' dosyasƒ± '{file_path}' yolunda bulunamadƒ±.")
    print("L√ºtfen CSV dosyasƒ±nƒ± Colab'a doƒüru ≈üekilde y√ºklediƒüinizden ve 'csv_file_name' deƒüi≈ükenini doƒüru ayarladƒ±ƒüƒ±nƒ±zdan emin olun.")
    exit()

df = pd.read_csv(file_path)

print(f"'{csv_file_name}' ba≈üarƒ±yla y√ºklendi. Toplam {len(df)} film bulundu.")
print("\nVeri Setinin ƒ∞lk 5 Satƒ±rƒ±:")
print(df.head())
print("\nADIM 1: Veri Seti Ke≈üfi Tamamlandƒ±.")

# --- ADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme ---
print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Ba≈ülƒ±yor (T√ºrler Basitle≈ütiriliyor)...")

df_filtered = df[['Title', 'IMDb Rating', 'Tags', 'Director', 'Stars', 'Votes']].copy()

df_filtered['Stars'].fillna('', inplace=True)

# T√ºrleri Basitle≈ütirme i√ßin e≈üleme s√∂zl√ºƒü√º
genre_mapping = {
    'action': ['action', 'action epic', 'gun fu', 'one-person army action', 'car action', 'kung fu', 'martial arts'],
    'adventure': ['adventure', 'adventure epic', 'desert adventure', 'animal adventure', 'space adventure', 'swashbuckler'],
    'comedy': ['comedy', 'romantic comedy', 'buddy comedy', 'sitcom', 'black comedy', 'satire', 'spoof', 'parody', 'slapstick', 'screwball comedy', 'dark comedy', 'body swap comedy'],
    'drama': ['drama', 'period drama', 'cop drama', 'legal drama', 'medical drama', 'teen drama', 'psychological drama', 'melodrama', 'historical drama', 'biography', 'romantic drama', 'showbiz drama', 'tragedy'],
    'thriller': ['thriller', 'crime thriller', 'spy thriller', 'psychological thriller', 'mystery thriller', 'political thriller', 'conspiracy thriller', 'erotic thriller', 'cyber thriller', 'suspense'],
    'sci-fi': ['sci-fi', 'space sci-fi', 'dystopian sci-fi', 'cyberpunk', 'alien invasion', 'mutant', 'robot', 'post-apocalyptic', 'time travel'],
    'fantasy': ['fantasy', 'dark fantasy', 'sword & sorcery', 'fairy tale', 'epic fantasy'],
    'horror': ['horror', 'slasher', 'supernatural horror', 'body horror', 'zombie', 'monster', 'vampire', 'werewolf', 'ghost'],
    'mystery': ['mystery', 'suspense mystery', 'cozy mystery', 'whodunnit', 'detective', 'police procedural'],
    'crime': ['crime', 'gangster', 'heist', 'mob', 'true crime'],
    'romance': ['romance', 'romantic comedy', 'romantic drama'],
    'animation': ['animation', 'adult animation', 'anime', 'computer animation', 'drawn animation', 'stop-motion animation'],
    'family': ['family', 'kids'],
    'western': ['western', 'classic western', 'neo-western'],
    'war': ['war', 'war drama'],
    'history': ['history', 'historical drama', 'biography'],
    'music': ['music', 'musical', 'classic musical', 'concert'],
    'documentary': ['documentary', 'docudrama', 'mockumentary']
}

reverse_genre_map = {}
for main_genre, sub_genres in genre_mapping.items():
    for sub_genre in sub_genres:
        reverse_genre_map[sub_genre] = main_genre

def map_to_main_genres(tag_list):
    main_genres = set()
    for tag in tag_list:
        main_genres.add(reverse_genre_map.get(tag, tag))
    return list(main_genres)

def clean_and_split(text_series):
    if isinstance(text_series, float) and pd.isna(text_series): # NaN kontrol√º
        return []

    # ensure it's a string before processing
    item = str(text_series)
    item = item.replace('"', '').replace("'", '').strip()
    split_items = [s.strip().lower() for s in item.split(',') if s.strip()]
    return split_items


df_filtered['Tags_cleaned_raw'] = df_filtered['Tags'].apply(clean_and_split)
df_filtered['Director_cleaned'] = df_filtered['Director'].apply(clean_and_split)
df_filtered['Stars_cleaned'] = df_filtered['Stars'].apply(clean_and_split)

df_filtered['Tags_cleaned'] = df_filtered['Tags_cleaned_raw'].apply(map_to_main_genres)


def convert_votes_to_numeric(votes_str):
    if isinstance(votes_str, str):
        votes_str = votes_str.replace(",", "") # Virg√ºlleri temizle
        if 'K' in votes_str:
            return float(votes_str.replace('K', '')) * 1000
        elif 'M' in votes_str:
            return float(votes_str.replace('M', '')) * 1_000_000
    return float(votes_str)

df_filtered['Votes_numeric'] = df_filtered['Votes'].apply(convert_votes_to_numeric)
df_filtered.drop('Votes', axis=1, inplace=True)

print("\nTemizlenmi≈ü Veri Setinin ƒ∞lk 5 Satƒ±rƒ± (Yeni S√ºtunlarla):")
print(df_filtered.head())
print("\nBasitle≈ütirilmi≈ü T√ºrlerin √ñrnekleri:")
print(df_filtered['Tags_cleaned'].head())
print("\nADIM 2: Veri Temizliƒüi ve √ñn ƒ∞≈üleme Tamamlandƒ±.")

# --- ADIM 3: Film √ñneri Sistemi Mantƒ±ƒüƒ±nƒ± Olu≈üturma ---
print("\nADIM 3: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuluyor...")

all_tags = sorted(list(set([tag for sublist in df_filtered['Tags_cleaned'] for tag in sublist])))
all_directors = sorted(list(set([director for sublist in df_filtered['Director_cleaned'] for director in sublist])))
all_stars = sorted(list(set([star for sublist in df_filtered['Stars_cleaned'] for star in sublist])))

def get_movie_recommendations(selected_tags, selected_directors, selected_stars, min_imdb_rating_slider, num_recommendations_slider):

    print(f"\n--- √ñneri ƒ∞steƒüi ---")
    print(f"Se√ßilen T√ºrler: {selected_tags}")
    print(f"Se√ßilen Y√∂netmenler: {selected_directors}")
    print(f"Se√ßilen Oyuncular: {selected_stars}")
    print(f"Minimum IMDb Puanƒ±: {min_imdb_rating_slider}")
    print(f"√ñneri Sayƒ±sƒ±: {num_recommendations_slider}")
    print(f"Ba≈ülangƒ±√ß DataFrame boyutu: {len(df_filtered)}")

    selected_tags = list(selected_tags) if selected_tags else []
    selected_directors = list(selected_directors) if selected_directors else []
    selected_stars = list(selected_stars) if selected_stars else []

    recommendations_df = df_filtered.copy()

    recommendations_df = recommendations_df[recommendations_df['IMDb Rating'] >= min_imdb_rating_slider]
    print(f"IMDb Puanƒ± filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    if selected_tags:
        recommendations_df = recommendations_df[
            recommendations_df['Tags_cleaned'].apply(lambda x: any(tag in x for tag in selected_tags))
        ]
        print(f"T√ºr filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    if selected_directors:
        recommendations_df = recommendations_df[
            recommendations_df['Director_cleaned'].apply(lambda x: any(director in x for director in selected_directors))
        ]
        print(f"Y√∂netmen filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    if selected_stars:
        recommendations_df = recommendations_df[
            recommendations_df['Stars_cleaned'].apply(lambda x: any(star in x for star in selected_stars))
        ]
        print(f"Oyuncu filtrelemesi sonrasƒ±: {len(recommendations_df)} film")

    recommendations_df = recommendations_df.sort_values(
        by=['IMDb Rating', 'Votes_numeric'],
        ascending=[False, False]
    ).reset_index(drop=True)

    top_recommendations = recommendations_df.head(num_recommendations_slider)

    if top_recommendations.empty:
        print("Kriterlere uygun film bulunamadƒ±.")
        return "√úzg√ºn√ºz, se√ßtiƒüiniz kriterlere uygun film bulunamadƒ±."
    else:
        print(f"Toplam {len(top_recommendations)} √∂neri bulundu.")
        recommendation_list = []
        for idx, row in top_recommendations.iterrows():
            directors_str = ", ".join([d.title() for d in row['Director_cleaned']])
            stars_str = ", ".join([s.title() for s in row['Stars_cleaned']])
            tags_str = ", ".join([t.title() for t in row['Tags_cleaned']])

            recommendation_list.append(
                f"**{row['Title']}** (IMDb: {row['IMDb Rating']:.1f} ‚≠êÔ∏è, Oylar: {int(row['Votes_numeric']):,} üó≥Ô∏è)"
                f"\n  Y√∂netmen: {directors_str if directors_str else 'Bilinmiyor'}"
                f"\n  Oyuncular: {stars_str if stars_str else 'Bilinmiyor'}"
                f"\n  T√ºrler: {tags_str if tags_str else 'Bilinmiyor'}"
                f"\n---"
            )
        return "\n".join(recommendation_list)

print("\nADIM 3: Film √ñneri Sistemi Mantƒ±ƒüƒ± Olu≈üturuldu.")

# --- ADIM 4: Gradio Web Aray√ºz√º Olu≈üturma ---
print("\nADIM 4: Gradio Web Aray√ºz√º Olu≈üturuluyor...")

with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown(
        """
        # üé¨ Film √ñneri Sistemi
        Favori film √∂zelliklerinizi se√ßin, y√ºksek IMDb puanƒ±na sahip filmleri ke≈üfedin!
        """
    )

    with gr.Row():
        with gr.Column():
            gr.Markdown("### Film √ñzelliklerini Se√ßin:")

            tags_input = gr.CheckboxGroup(
                label="Film T√ºrleri",
                choices=all_tags,
                value=['action', 'drama'],
                interactive=True
            )

            directors_input = gr.Dropdown(
                label="Y√∂netmenler",
                choices=all_directors,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            stars_input = gr.Dropdown(
                label="Oyuncular",
                choices=all_stars,
                multiselect=True,
                allow_custom_value=False,
                interactive=True
            )

            min_imdb_rating_slider = gr.Slider(
                minimum=df_filtered['IMDb Rating'].min(),
                maximum=df_filtered['IMDb Rating'].max(),
                step=0.1,
                value=7.6,
                label="Minimum IMDb Puanƒ±"
            )

            num_recommendations_slider = gr.Slider(
                minimum=1,
                maximum=20,
                step=1,
                value=10,
                label="√ñneri Sayƒ±sƒ±"
            )

            recommend_btn = gr.Button("üöÄ Film √ñnerilerini Getir", variant="primary", size="lg")

        with gr.Column():
            gr.Markdown("### √ñnerilen Filmler:")
            output_text = gr.Markdown(label="√ñnerileriniz burada listelenecektir.", value="Hen√ºz bir √∂neri yapƒ±lmadƒ±. √ñzellikleri se√ßip butona tƒ±klayƒ±n!")

    gr.Examples(
        examples=[
            [['action'], [], [], 7.6, 5],
            [['comedy', 'drama'], [], [], 7.8, 3],
            [[], ['christopher nolan'], [], 8.0, 5],
            [[], [], ['leonardo dicaprio'], 7.8, 3],
        ],
        inputs=[tags_input, directors_input, stars_input, min_imdb_rating_slider, num_recommendations_slider],
        outputs=output_text,
        fn=get_movie_recommendations,
        label="√ñrnek √ñnerileri Deneyin"
    )

    gr.Markdown(
        """
        ---
        ### ‚ÑπÔ∏è Nasƒ±l Kullanƒ±lƒ±r?
        1.  **Film T√ºrleri, Y√∂netmenler ve Oyuncular** b√∂l√ºmlerinden istediƒüiniz filtreleri se√ßin (birden fazla se√ßim yapabilirsiniz).
        2.  **Minimum IMDb Puanƒ±** ve **√ñneri Sayƒ±sƒ±** √ßubuklarƒ±nƒ± ayarlayƒ±n.
        3.  **"Film √ñnerilerini Getir"** butonuna tƒ±klayƒ±n.
        4.  √ñneriler saƒü panelde g√∂r√ºnecektir!
        """
    )

demo.launch(share=True)
print("\nADIM 4: Gradio Web Aray√ºz√º Ba≈ülatƒ±ldƒ±.")
```
